```python
# =============================
# Kaggle-ready Captum Explainer - UPDATED
# Integrated with new preprocessing pipeline + Emoji visualization
# =============================

import os
import numpy as np
import pandas as pd
import torch
import joblib
import re
import emoji
import regex
import warnings
warnings.filterwarnings("ignore")
from typing import Dict, List, Tuple, Optional
from tqdm import tqdm

from transformers import AutoTokenizer, AutoModelForSequenceClassification
from huggingface_hub import hf_hub_download
from matplotlib import pyplot as plt, cm
from matplotlib.font_manager import FontProperties, fontManager
import matplotlib.colors as mcolors

# Captum
from captum.attr import LayerIntegratedGradients

# Preprocessing imports
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
from deep_translator import GoogleTranslator
from functools import lru_cache
import logging
from typing import Literal

logging.basicConfig(level=logging.WARNING)
logger = logging.getLogger(__name__)

try:
    from indic_transliteration import sanscript
    from indic_transliteration.sanscript import transliterate
    TRANSLITERATION_AVAILABLE = True
except ImportError:
    TRANSLITERATION_AVAILABLE = False
    logger.warning("indic_transliteration not available. Transliteration disabled.")


# ----------------------------
# CONFIGURATION
# ----------------------------
LOCAL_MODEL_PATH = '/kaggle/input/xlm-roberta-nepali-hate'
HF_MODEL_ID = "UDHOV/xlm-roberta-large-nepali-hate-classification"
TEST_FILE = '/kaggle/input/nepalihatee/test_processed.json'
OUT_DIR = '/kaggle/working/captum_explanations'
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
MAX_LENGTH = 256
os.makedirs(OUT_DIR, exist_ok=True)
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)


# ============================ FONT CONFIG ==================================
FONT_PATH = '/kaggle/input/fontss/Kalimati.ttf'

def load_nepali_font(font_path):
    if not os.path.exists(font_path):
        print(f"âš ï¸ Nepali font not found: {font_path}")
        return None
    try:
        fontManager.addfont(font_path)
        fp = FontProperties(fname=font_path)
        print(f"âœ“ Loaded Nepali font: {fp.get_name()}")
        return fp
    except Exception as e:
        print(f"âš ï¸ Could not load font: {e}")
        return None

NEPALI_FONT = load_nepali_font(FONT_PATH)
plt.rcParams["axes.unicode_minus"] = False


# ============================================================================
# ENHANCED EMOJI MAPPING (Same as LIME/SHAP)
# ============================================================================

EMOJI_TO_NEPALI = {
    # ===== Laughter & Joy (Positive) =====
    'ğŸ˜‚': 'à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ¤£': 'à¤ à¥‚à¤²à¥‹ à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ˜€': 'à¤–à¥à¤¶à¥€', 'ğŸ˜': 'à¤–à¥à¤¶à¥€', 'ğŸ˜ƒ': 'à¤–à¥à¤¶à¥€',
    'ğŸ˜„': 'à¤–à¥à¤¶à¥€', 'ğŸ˜…': 'à¤¨à¤°à¥à¤­à¤¸ à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ˜†': 'à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ˜Š': 'à¤®à¥à¤¸à¥à¤•à¤¾à¤¨', 'â˜ºï¸': 'à¤®à¥à¤¸à¥à¤•à¤¾à¤¨',
    'ğŸ˜‰': 'à¤†à¤à¤–à¤¾ à¤à¤¿à¤®à¥à¤•à¤¾à¤‰à¤¨à¥‡', 'ğŸ™‚': 'à¤®à¥à¤¸à¥à¤•à¤¾à¤¨', 'ğŸ™ƒ': 'à¤‰à¤²à¥à¤Ÿà¥‹ à¤®à¥à¤¸à¥à¤•à¤¾à¤¨', 'ğŸ˜Œ': 'à¤¶à¤¾à¤¨à¥à¤¤',
    'ğŸ˜': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ¥°': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ˜˜': 'à¤šà¥à¤®à¥à¤¬à¤¨', 'ğŸ˜—': 'à¤šà¥à¤®à¥à¤¬à¤¨', 'ğŸ˜™': 'à¤šà¥à¤®à¥à¤¬à¤¨', 'ğŸ˜š': 'à¤šà¥à¤®à¥à¤¬à¤¨',
    'ğŸ¤—': 'à¤…à¤à¤—à¤¾à¤²à¥‹', 'ğŸ¤©': 'à¤šà¤•à¤¿à¤¤', 'ğŸ¥³': 'à¤‰à¤¤à¥à¤¸à¤µ', 'ğŸ¤¤': 'à¤²à¤¾à¤²à¤¸à¤¾',
    
    # ===== Mocking & Sarcasm (Hate Indicator) =====
    'ğŸ˜': 'à¤µà¥à¤¯à¤‚à¤—à¥à¤¯', 'ğŸ˜œ': 'à¤œà¤¿à¤¬à¥à¤°à¥‹ à¤¦à¥‡à¤–à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜': 'à¤œà¤¿à¤¬à¥à¤°à¥‹ à¤¦à¥‡à¤–à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜›': 'à¤œà¤¿à¤¬à¥à¤°à¥‹',
    'ğŸ™„': 'à¤†à¤à¤–à¤¾ à¤˜à¥à¤®à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜¤': 'à¤¨à¤¿à¤°à¤¾à¤¶', 'ğŸ˜‘': 'à¤…à¤­à¤¿à¤µà¥à¤¯à¤•à¥à¤¤à¤¿à¤¹à¥€à¤¨', 'ğŸ˜': 'à¤¤à¤Ÿà¤¸à¥à¤¥',
    'ğŸ˜¬': 'à¤¤à¤¨à¤¾à¤µ', 'ğŸ¤¨': 'à¤¶à¤‚à¤•à¤¾à¤¸à¥à¤ªà¤¦', 'ğŸ¤«': 'à¤šà¥à¤ªà¤šà¤¾à¤ª',
    'ğŸ¤­': 'à¤¹à¤¾à¤¤ à¤›à¥‹à¤ªà¥à¤¨à¥‡', 'ğŸ¤¥': 'à¤à¥‚à¤ ', 'ğŸ˜¶': 'à¤®à¥Œà¤¨',
    
    # ===== Anger & Hate (CRITICAL for Detection) =====
    'ğŸ˜ ': 'à¤°à¤¿à¤¸', 'ğŸ˜¡': 'à¤ à¥‚à¤²à¥‹ à¤°à¤¿à¤¸', 'ğŸ¤¬': 'à¤—à¤¾à¤²à¥€', 'ğŸ˜ˆ': 'à¤–à¤°à¤¾à¤¬', 'ğŸ‘¿': 'à¤–à¤°à¤¾à¤¬',
    'ğŸ’¢': 'à¤•à¥à¤°à¥‹à¤§', 'ğŸ”ª': 'à¤¹à¤¿à¤‚à¤¸à¤¾', 'ğŸ’£': 'à¤¹à¤¿à¤‚à¤¸à¤¾', 'ğŸ—¡ï¸': 'à¤¤à¤°à¤µà¤¾à¤°', 'âš”ï¸': 'à¤¯à¥à¤¦à¥à¤§',
    'ğŸ’¥': 'à¤µà¤¿à¤¸à¥à¤«à¥‹à¤Ÿ', 'ğŸ”«': 'à¤¬à¤¨à¥à¤¦à¥à¤•', 'ğŸ§¨': 'à¤µà¤¿à¤¸à¥à¤«à¥‹à¤Ÿà¤•', 'â˜ ï¸': 'à¤®à¥ƒà¤¤à¥à¤¯à¥', 'ğŸ’€': 'à¤–à¥‹à¤ªà¤¡à¥€',
    'ğŸ‘¹': 'à¤°à¤¾à¤•à¥à¤·à¤¸', 'ğŸ‘º': 'à¤¦à¤¾à¤¨à¤µ', 'ğŸ¤¡': 'à¤œà¥‹à¤•à¤°', 'ğŸ–¤': 'à¤•à¤¾à¤²à¥‹ à¤®à¤¨',
    'ğŸ˜¾': 'à¤°à¤¿à¤¸à¤¾à¤à¤•à¥‹', 'ğŸ‘Š': 'à¤®à¥à¤•à¥à¤•à¤¾', 'âœŠ': 'à¤®à¥à¤•à¥à¤•à¤¾',
    
    # ===== Offensive Gestures (Strong Hate Signal) =====
    'ğŸ–•': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ‘': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ»': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¼': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•',
    'ğŸ‘ğŸ½': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¾': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¿': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•',
    'ğŸ–•ğŸ»': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ¼': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ½': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ¾': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ¿': 'à¤…à¤ªà¤®à¤¾à¤¨',
    
    # ===== Sadness & Crying =====
    'ğŸ˜­': 'à¤°à¥à¤µà¤¾à¤‡', 'ğŸ˜¢': 'à¤°à¥à¤µà¤¾à¤‡', 'ğŸ˜¿': 'à¤°à¥à¤µà¤¾à¤‡', 'ğŸ˜”': 'à¤‰à¤¦à¤¾à¤¸', 'ğŸ˜': 'à¤‰à¤¦à¤¾à¤¸',
    'ğŸ˜’': 'à¤‰à¤¦à¤¾à¤¸', 'ğŸ˜“': 'à¤šà¤¿à¤¨à¥à¤¤à¤¿à¤¤', 'ğŸ˜Ÿ': 'à¤šà¤¿à¤¨à¥à¤¤à¤¿à¤¤', 'ğŸ˜•': 'à¤…à¤²à¤®à¤²à¤¿à¤à¤•à¥‹',
    'ğŸ™': 'à¤¤à¤²à¥à¤²à¥‹ à¤®à¥à¤–', 'â˜¹ï¸': 'à¤¦à¥à¤ƒà¤–à¥€', 'ğŸ˜©': 'à¤¥à¤•à¤¿à¤¤', 'ğŸ˜«': 'à¤¥à¤•à¤¿à¤¤',
    'ğŸ˜–': 'à¤­à¥à¤°à¤®à¤¿à¤¤', 'ğŸ˜£': 'à¤…à¤¡à¤¿à¤—', 'ğŸ˜¥': 'à¤¨à¤¿à¤°à¤¾à¤¶', 'ğŸ¥º': 'à¤¬à¤¿à¤¨à¥à¤¤à¥€',
    
    # ===== Fear & Shock =====
    'ğŸ˜¨': 'à¤¡à¤°', 'ğŸ˜°': 'à¤šà¤¿à¤¨à¥à¤¤à¤¿à¤¤ à¤ªà¤¸à¤¿à¤¨à¤¾', 'ğŸ˜±': 'à¤šà¤¿à¤šà¥à¤¯à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜³': 'à¤²à¤œà¤¾à¤‰à¤¨à¥‡',
    'ğŸ¤¯': 'à¤®à¤¨ à¤‰à¤¡à¥‡à¤•à¥‹', 'ğŸ˜µ': 'à¤šà¤•à¥à¤•à¤°', 'ğŸ˜²': 'à¤šà¤•à¤¿à¤¤', 'ğŸ˜¯': 'à¤›à¤•à¥à¤•',
    
    # ===== Disgust & Contempt =====
    'ğŸ¤¢': 'à¤¬à¤¾à¤¨à¥à¤¤à¤¾', 'ğŸ¤®': 'à¤¬à¤¾à¤¨à¥à¤¤à¤¾', 'ğŸ¤§': 'à¤¹à¤¾à¤šà¥à¤›à¥à¤¯à¥‚à¤', 'ğŸ˜·': 'à¤¬à¤¿à¤°à¤¾à¤®à¥€',
    'ğŸ¤’': 'à¤œà¥à¤µà¤°à¥‹', 'ğŸ¤•': 'à¤˜à¤¾à¤‡à¤¤à¥‡', 'ğŸ¥´': 'à¤®à¤¾à¤¤à¥à¤¤à¤¿à¤à¤•à¥‹', 'ğŸ˜ª': 'à¤¨à¤¿à¤¦à¥à¤°à¤¾',
    
    # ===== Positive Gestures & Symbols =====
    'ğŸ‘': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ»': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¼': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 
    'ğŸ‘ğŸ½': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¾': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¿': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•',
    'ğŸ‘': 'à¤¤à¤¾à¤²à¤¿', 'ğŸ™Œ': 'à¤‰à¤¤à¥à¤¸à¤µ', 'ğŸ‘Œ': 'à¤ à¥€à¤• à¤›', 'ğŸ¤': 'à¤¹à¤¾à¤¤ à¤®à¤¿à¤²à¤¾à¤‰à¤¨à¥',
    'ğŸ™': 'à¤¨à¤®à¤¸à¥à¤•à¤¾à¤°', 'ğŸ¤²': 'à¤ªà¥à¤°à¤¾à¤°à¥à¤¥à¤¨à¤¾', 'ğŸ’ª': 'à¤¶à¤•à¥à¤¤à¤¿', 'âœŒï¸': 'à¤¶à¤¾à¤¨à¥à¤¤à¤¿',
    
    # ===== Hearts & Love =====
    'â¤ï¸': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ§¡': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’›': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’š': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’™': 'à¤®à¤¾à¤¯à¤¾',
    'ğŸ’œ': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ¤': 'à¤¸à¥‡à¤¤à¥‹ à¤®à¤¨', 'ğŸ¤': 'à¤–à¥ˆà¤°à¥‹ à¤®à¤¨',
    'â£ï¸': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’•': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’“': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’—': 'à¤®à¤¾à¤¯à¤¾',
    'ğŸ’–': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’˜': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’”': 'à¤Ÿà¥à¤Ÿà¥‡à¤•à¥‹ à¤®à¤¨',
    
    # ===== Symbols & Objects =====
    'ğŸ”¥': 'à¤†à¤—à¥‹', 'ğŸ’¯': 'à¤ªà¥‚à¤°à¥à¤£', 'ğŸ’¨': 'à¤¹à¤¾à¤µà¤¾', 'ğŸ’«': 'à¤šà¤®à¤•',
    'â­': 'à¤¤à¤¾à¤°à¤¾', 'âœ¨': 'à¤šà¤®à¤•', 'ğŸŒŸ': 'à¤šà¤®à¥à¤•à¤¿à¤²à¥‹ à¤¤à¤¾à¤°à¤¾',
    'ğŸš«': 'à¤¨à¤¿à¤·à¥‡à¤§', 'â›”': 'à¤ªà¥à¤°à¤µà¥‡à¤¶ à¤¨à¤¿à¤·à¥‡à¤§', 'âŒ': 'à¤°à¤¦à¥à¤¦', 'â': 'à¤—à¤²à¤¤',
    
    # ===== People & Relationships =====
    'ğŸ‘«': 'à¤œà¥‹à¤¡à¥€', 'ğŸ‘¬': 'à¤ªà¥à¤°à¥à¤· à¤œà¥‹à¤¡à¥€', 'ğŸ‘­': 'à¤®à¤¹à¤¿à¤²à¤¾ à¤œà¥‹à¤¡à¥€', 'ğŸ‘¨â€ğŸ‘©â€ğŸ‘§â€ğŸ‘¦': 'à¤ªà¤°à¤¿à¤µà¤¾à¤°',
    'ğŸ‘¶': 'à¤¬à¤šà¥à¤šà¤¾', 'ğŸ‘¦': 'à¤•à¥‡à¤Ÿà¤¾', 'ğŸ‘§': 'à¤•à¥‡à¤Ÿà¥€', 'ğŸ‘¨': 'à¤ªà¥à¤°à¥à¤·', 'ğŸ‘©': 'à¤®à¤¹à¤¿à¤²à¤¾',
    'ğŸ‘´': 'à¤¬à¥‚à¤¢à¥‹', 'ğŸ‘µ': 'à¤¬à¥‚à¤¢à¥€', 'ğŸ§’': 'à¤¬à¤¾à¤²à¤•', 'ğŸ‘±': 'à¤—à¥‹à¤°à¥‹', 'ğŸ§”': 'à¤¦à¤¾à¤¹à¥à¤°à¥€',
    
    # ===== Country & Culture =====
    'ğŸ‡³ğŸ‡µ': 'à¤¨à¥‡à¤ªà¤¾à¤²', 'ğŸ‡®ğŸ‡³': 'à¤­à¤¾à¤°à¤¤', 'ğŸ‡µğŸ‡°': 'à¤ªà¤¾à¤•à¤¿à¤¸à¥à¤¤à¤¾à¤¨', 'ğŸ‡§ğŸ‡©': 'à¤¬à¤‚à¤—à¤²à¤¾à¤¦à¥‡à¤¶',
    'ğŸ‡¨ğŸ‡³': 'à¤šà¥€à¤¨', 'ğŸ‡ºğŸ‡¸': 'à¤…à¤®à¥‡à¤°à¤¿à¤•à¤¾', 'ğŸ´': 'à¤à¤£à¥à¤¡à¤¾',
    
    # ===== Animals (sometimes used in hate) =====
    'ğŸ•': 'à¤•à¥à¤•à¥à¤°', 'ğŸ–': 'à¤¸à¥à¤à¤—à¥à¤°', 'ğŸ€': 'à¤®à¥à¤¸à¤¾', 'ğŸ': 'à¤¸à¤°à¥à¤ª', 'ğŸ¦‚': 'à¤¬à¤¿à¤šà¥à¤›à¥€',
    'ğŸ’': 'à¤¬à¤¾à¤à¤¦à¤°', 'ğŸµ': 'à¤¬à¤¾à¤à¤¦à¤° à¤…à¤¨à¥à¤¹à¤¾à¤°', 'ğŸ¦': 'à¤—à¥‹à¤°à¤¿à¤²à¥à¤²à¤¾', 'ğŸ—': 'à¤œà¤™à¥à¤—à¤²à¥€ à¤¸à¥à¤à¤—à¥à¤°',
    
    # ===== Other Common =====
    'ğŸ¤”': 'à¤¸à¥‹à¤š', 'ğŸ§': 'à¤…à¤¨à¥à¤¸à¤¨à¥à¤§à¤¾à¤¨', 'ğŸ˜´': 'à¤¸à¥à¤¤à¥à¤¨à¥‡', 'ğŸ’©': 'à¤®à¤²',
    'ğŸ‘»': 'à¤­à¥‚à¤¤', 'ğŸ¤–': 'à¤°à¥‹à¤¬à¥‹à¤Ÿ', 'ğŸ‘½': 'à¤µà¤¿à¤¦à¥‡à¤¶à¥€', 'ğŸ­': 'à¤®à¥à¤–à¥Œà¤Ÿà¤¾',
}

# Emoji categories for feature extraction
HATE_RELATED_EMOJIS = {
    'ğŸ˜ ', 'ğŸ˜¡', 'ğŸ¤¬', 'ğŸ˜ˆ', 'ğŸ‘¿', 'ğŸ’¢', 'ğŸ‘Š', 'âœŠ',
    'ğŸ”ª', 'ğŸ’£', 'ğŸ—¡ï¸', 'âš”ï¸', 'ğŸ’¥', 'ğŸ”«', 'ğŸ§¨', 'â˜ ï¸', 'ğŸ’€',
    'ğŸ–•', 'ğŸ–•ğŸ»', 'ğŸ–•ğŸ¼', 'ğŸ–•ğŸ½', 'ğŸ–•ğŸ¾', 'ğŸ–•ğŸ¿',
    'ğŸ‘', 'ğŸ‘ğŸ»', 'ğŸ‘ğŸ¼', 'ğŸ‘ğŸ½', 'ğŸ‘ğŸ¾', 'ğŸ‘ğŸ¿',
    'ğŸ‘¹', 'ğŸ‘º', 'ğŸ¤¡', 'ğŸ–¤', 'ğŸ’”',
    'ğŸ•', 'ğŸ–', 'ğŸ€', 'ğŸ', 'ğŸ¦‚', 'ğŸ’', 'ğŸµ', 'ğŸ¦', 'ğŸ—',
    'ğŸ’©', 'ğŸ˜¾',
}

MOCKERY_EMOJIS = {
    'ğŸ˜', 'ğŸ˜œ', 'ğŸ˜', 'ğŸ˜›', 'ğŸ™„', 'ğŸ˜¤', 'ğŸ™ƒ',
    'ğŸ˜‘', 'ğŸ˜', 'ğŸ˜¬', 'ğŸ¤¨', 'ğŸ¤«', 'ğŸ¤­', 'ğŸ¤¥',
    'ğŸ¤¡', 'ğŸ‘»', 'ğŸ­',
}

POSITIVE_EMOJIS = {
    'ğŸ˜Š', 'ğŸ˜€', 'ğŸ˜', 'ğŸ˜ƒ', 'ğŸ˜„', 'â˜ºï¸', 'ğŸ™‚', 'ğŸ˜Œ', 'ğŸ¥°', 'ğŸ˜',
    'â¤ï¸', 'ğŸ§¡', 'ğŸ’›', 'ğŸ’š', 'ğŸ’™', 'ğŸ’œ', 'ğŸ¤', 'ğŸ¤',
    'ğŸ’•', 'ğŸ’', 'ğŸ’“', 'ğŸ’—', 'ğŸ’–', 'ğŸ’˜', 'ğŸ’', 'â£ï¸',
    'ğŸ‘', 'ğŸ‘ğŸ»', 'ğŸ‘ğŸ¼', 'ğŸ‘ğŸ½', 'ğŸ‘ğŸ¾', 'ğŸ‘ğŸ¿',
    'ğŸ™', 'ğŸ‘', 'ğŸ™Œ', 'ğŸ‘Œ', 'ğŸ¤', 'âœŒï¸',
    'ğŸ¥³', 'ğŸ‰', 'ğŸŠ', 'â­', 'âœ¨', 'ğŸŒŸ',
}
SADNESS_EMOJIS = {
    'ğŸ˜­', 'ğŸ˜¢', 'ğŸ˜¿', 'ğŸ˜”', 'ğŸ˜', 'ğŸ˜’', 'ğŸ˜“', 'ğŸ˜Ÿ', 'ğŸ˜•',
    'ğŸ™', 'â˜¹ï¸', 'ğŸ˜©', 'ğŸ˜«', 'ğŸ˜–', 'ğŸ˜£', 'ğŸ˜¥', 'ğŸ¥º',
}

FEAR_EMOJIS = {
    'ğŸ˜¨', 'ğŸ˜°', 'ğŸ˜±', 'ğŸ˜³', 'ğŸ¤¯', 'ğŸ˜µ', 'ğŸ˜²', 'ğŸ˜¯',
}

DISGUST_EMOJIS = {
    'ğŸ¤¢', 'ğŸ¤®', 'ğŸ¤§', 'ğŸ˜·', 'ğŸ¤’', 'ğŸ¤•', 'ğŸ¥´',
}


# ============================================================================
# EMOJI INTELLIGENCE - FEATURE EXTRACTION
# ============================================================================

def extract_emoji_features(text: str) -> Dict[str, int]:
    """Extract comprehensive emoji-based semantic features"""
    emojis_found = [c for c in text if c in emoji.EMOJI_DATA]
    
    hate_count = sum(1 for e in emojis_found if e in HATE_RELATED_EMOJIS)
    mockery_count = sum(1 for e in emojis_found if e in MOCKERY_EMOJIS)
    positive_count = sum(1 for e in emojis_found if e in POSITIVE_EMOJIS)
    sadness_count = sum(1 for e in emojis_found if e in SADNESS_EMOJIS)
    fear_count = sum(1 for e in emojis_found if e in FEAR_EMOJIS)
    disgust_count = sum(1 for e in emojis_found if e in DISGUST_EMOJIS)
    
    return {
        'has_hate_emoji': 1 if hate_count > 0 else 0,
        'has_mockery_emoji': 1 if mockery_count > 0 else 0,
        'has_positive_emoji': 1 if positive_count > 0 else 0,
        'has_sadness_emoji': 1 if sadness_count > 0 else 0,
        'has_fear_emoji': 1 if fear_count > 0 else 0,
        'has_disgust_emoji': 1 if disgust_count > 0 else 0,
        'hate_emoji_count': hate_count,
        'mockery_emoji_count': mockery_count,
        'positive_emoji_count': positive_count,
        'sadness_emoji_count': sadness_count,
        'fear_emoji_count': fear_count,
        'disgust_emoji_count': disgust_count,
        'total_emoji_count': len(emojis_found),
        'hate_to_positive_ratio': hate_count / max(positive_count, 1),
        'has_mixed_sentiment': 1 if (hate_count > 0 and positive_count > 0) else 0,
    }


# ============================================================================
# SCRIPT DETECTION (IGNORES EMOJIS)
# ============================================================================

ScriptType = Literal["devanagari", "romanized_nepali", "english", "mixed", "other"]


def remove_emojis_for_detection(text: str) -> str:
    """Remove emojis temporarily for script detection"""
    return emoji.replace_emoji(text, replace="")


def detect_script_type(text: str) -> Tuple[ScriptType, dict]:
    """Detect the dominant script type IGNORING emojis"""
    if not text or not text.strip():
        return "other", {"confidence": 0.0, "reason": "empty_text"}
    
    text_no_emoji = remove_emojis_for_detection(text)
    
    if not text_no_emoji.strip():
        return "other", {"confidence": 0.5, "reason": "emoji_only"}
    
    letters = regex.findall(r"\p{L}", text_no_emoji)
    letter_count = len(letters)
    
    if letter_count == 0:
        return "other", {"confidence": 0.0, "reason": "no_letters"}
    
    devanagari_chars = regex.findall(r"\p{Devanagari}", text_no_emoji)
    dev_count = len(devanagari_chars)
    dev_ratio = dev_count / letter_count
    
    latin_chars = regex.findall(r"[a-zA-Z]", text_no_emoji)
    latin_count = len(latin_chars)
    latin_ratio = latin_count / letter_count
    
    romanized_nepali_patterns = [
        r'\b[xX]u\b', r'\b[xX]um?\b', r'\bhajur\b', r'\bdai\b', r'\bbhai\b',
        r'\bko\b', r'\bki\b', r'\bho\b', r'\bxa\b', r'\bxaina\b',
    ]
    
    romanized_indicators = sum(1 for pattern in romanized_nepali_patterns 
                               if re.search(pattern, text_no_emoji, re.IGNORECASE))
    
    romanized_score = 0.0
    if latin_ratio > 0.5 and dev_ratio < 0.3:
        if romanized_indicators > 0:
            romanized_score = min(0.5 + (romanized_indicators * 0.15), 0.95)
        else:
            romanized_patterns = re.findall(r'\b\w*[aeiou](?:h)?\b', text_no_emoji.lower())
            if any(word.endswith(('xu', 'ro', 'no', 'lo', 'ko', 'ho')) 
                   for word in romanized_patterns):
                romanized_score = 0.4
            else:
                romanized_score = 0.3
    
    english_indicators = [
        'the', 'a', 'an', 'this', 'that', 'is', 'am', 'are', 'was', 'were',
        'i', 'you', 'he', 'she', 'it', 'we', 'they', 'what', 'when', 'where',
    ]
    english_words = [w.lower() for w in re.findall(r'\b\w+\b', text_no_emoji)]
    english_count = sum(1 for w in english_words if w in english_indicators)
    english_ratio = english_count / len(english_words) if english_words else 0
    
    details = {
        "devanagari_count": dev_count,
        "devanagari_ratio": dev_ratio,
        "latin_count": latin_count,
        "latin_ratio": latin_ratio,
        "romanized_indicators": romanized_indicators,
        "english_ratio": english_ratio,
        "letter_count": letter_count
    }

    if dev_ratio >= 0.8:
        return "devanagari", {**details, "confidence": dev_ratio, "reason": "dominant_devanagari"}
    elif dev_ratio >= 0.4:
        return "mixed", {**details, "confidence": 0.7, "reason": "mixed_with_devanagari"}
    elif romanized_score > 0.5 and dev_ratio < 0.2:
        return "romanized_nepali", {**details, "confidence": romanized_score, "reason": "romanized_nepali_detected"}
    elif english_ratio > 0.2 and romanized_score < 0.4:
        return "english", {**details, "confidence": min(english_ratio + 0.3, 0.9), "reason": "english_detected"}
    elif latin_ratio > 0.5 and romanized_score > 0.3:
        return "romanized_nepali", {**details, "confidence": romanized_score, "reason": "likely_romanized_nepali"}
    elif latin_ratio > 0.8:
        if english_ratio > 0.1:
            return "english", {**details, "confidence": 0.6, "reason": "likely_english"}
        else:
            return "romanized_nepali", {**details, "confidence": 0.5, "reason": "ambiguous_latin_script"}
    else:
        return "other", {**details, "confidence": 0.3, "reason": "insufficient_indicators"}


# ============================================================================
# CACHED TRANSLATOR
# ============================================================================

class CachedNepaliTranslator:
    """Production-ready translator with caching"""
    
    def __init__(self, cache_size: int = 2000):
        self.translator = GoogleTranslator(source='en', target='ne')
        self.cache_size = cache_size
        self._translate_cached = lru_cache(maxsize=cache_size)(self._translate_single)
    
    def _translate_single(self, text: str) -> str:
        if not text or not text.strip():
            return ""
        try:
            result = self.translator.translate(text.strip())
            return result if result else text
        except Exception as e:
            logger.error(f"Translation failed: {str(e)}")
            raise
    
    def translate(self, text: str, fallback_to_original: bool = True) -> str:
        if not text or not text.strip():
            return ""
        try:
            return self._translate_cached(text.strip())
        except Exception as e:
            if fallback_to_original:
                logger.warning(f"Translation failed, using original: {str(e)}")
                return text
            raise


# ============================================================================
# SCRIPT-SPECIFIC PROCESSING
# ============================================================================

def transliterate_romanized_nepali(text: str) -> str:
    """Transliterate Romanized Nepali to Devanagari using ITRANS"""
    if not TRANSLITERATION_AVAILABLE:
        logger.warning("Transliteration not available")
        return text
    
    try:
        result = transliterate(text, sanscript.ITRANS, sanscript.DEVANAGARI)
        return result if result else text
    except Exception as e:
        logger.warning(f"Transliteration failed: {e}")
        return text


def translate_latin_spans(text: str, translator: CachedNepaliTranslator) -> str:
    """Translate ONLY Latin word spans inside a Devanagari-dominant sentence"""
    def repl(match):
        latin_text = match.group(0)
        translated = translator.translate(latin_text, fallback_to_original=True)
        return f" {translated} "
    
    return re.sub(r"[A-Za-z][A-Za-z\s]{2,}", repl, text)


# ============================================================================
# EMOJI PROCESSING
# ============================================================================

def replace_emojis_semantic(text: str, preserve_spacing: bool = True) -> str:
    """Replace emojis with Nepali text"""
    for emoji_char, nepali_text in EMOJI_TO_NEPALI.items():
        if preserve_spacing:
            text = text.replace(emoji_char, f" {nepali_text} ")
        else:
            text = text.replace(emoji_char, nepali_text)
    
    text = emoji.replace_emoji(text, replace=" ")
    return text


# ============================================================================
# NORMALIZATION
# ============================================================================

DIRGHIKARAN_MAP = {
    "\u200d": "",
    "\u200c": "",
    "à¥¤": ".",
    "à¥¥": ".",
}


def normalize_devanagari(text: str) -> str:
    """Normalize Devanagari-specific characters"""
    for k, v in DIRGHIKARAN_MAP.items():
        text = text.replace(k, v)
    return text


def clean_text_basic(text: str) -> str:
    """Basic cleaning: URLs, mentions, whitespace"""
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#(\w+)", r"\1", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


# ============================================================================
# MAIN PREPROCESSING PIPELINE
# ============================================================================

class HateSpeechPreprocessor:
    """Unified preprocessing pipeline for Nepali hate speech classification"""

    def __init__(
        self,
        model_type: Literal["xlmr", "mbert", "nepalibert"] = "xlmr",
        translate_english: bool = True,
        cache_size: int = 2000
    ):
        self.model_type = model_type
        self.translate_english = translate_english
        self.translator = CachedNepaliTranslator(cache_size) if translate_english else None

    def preprocess(self, text: str, verbose: bool = False) -> Tuple[str, Dict[str, int]]:
        """
        Main preprocessing pipeline
        
        Returns:
            Tuple of (preprocessed_text, emoji_features)
        """
        if not isinstance(text, str) or not text.strip():
            return "", {
                'has_hate_emoji': 0, 'has_mockery_emoji': 0, 'has_positive_emoji': 0,
                'has_sadness_emoji': 0, 'has_fear_emoji': 0, 'has_disgust_emoji': 0,
                'hate_emoji_count': 0, 'mockery_emoji_count': 0, 'positive_emoji_count': 0,
                'sadness_emoji_count': 0, 'fear_emoji_count': 0, 'disgust_emoji_count': 0,
                'total_emoji_count': 0, 'hate_to_positive_ratio': 0.0, 'has_mixed_sentiment': 0
            }


        original_text = text
        
        if verbose:
            print(f"ğŸ“¥ Original: {original_text}")
            
        # STEP 1: EXTRACT EMOJI FEATURES
        emoji_features = extract_emoji_features(original_text)

        # STEP 2: DETECT SCRIPT TYPE
        script_type, _ = detect_script_type(text)

        # STEP 3: BASIC CLEANING
        text = clean_text_basic(text)

        # STEP 4: SCRIPT-SPECIFIC PROCESSING
        if script_type == "devanagari":
            processed = text
            if self.translate_english:
                processed = translate_latin_spans(processed, self.translator)
        elif script_type == "romanized_nepali":
            processed = transliterate_romanized_nepali(text)
        elif script_type == "english":
            if self.translate_english:
                processed = self.translator.translate(text, fallback_to_original=True)
            else:
                processed = text
        elif script_type == "mixed":
            processed = transliterate_romanized_nepali(text)
            if self.translate_english:
                processed = translate_latin_spans(processed, self.translator)
        else:
            processed = text

        # STEP 5: EMOJI REPLACEMENT
        processed = replace_emojis_semantic(processed)

        # STEP 6: NORMALIZATION
        final = normalize_devanagari(processed)
        final = re.sub(r"\s+", " ", final).strip()

        return final, emoji_features


# ============================================================================
# TOKEN ALIGNMENT FOR EMOJI VISUALIZATION
# ============================================================================

def create_display_tokens_from_subwords(
    original_text: str,
    preprocessed_text: str,
    tokenizer_tokens: List[str],
    remove_special: bool = True
) -> List[str]:
    """
    Create display tokens that preserve emojis from original text
    Maps preprocessed tokens (with emoji translations) back to original tokens (with actual emojis)
    
    Strategy:
    1. Original: "à¤¤à¥‡à¤°à¥€ à¤•à¤¸à¤® ğŸ˜€ğŸ˜€" 
    2. Preprocessed: "à¤¤à¥‡à¤°à¥€ à¤•à¤¸à¤® à¤–à¥à¤¶à¥€ à¤–à¥à¤¶à¥€"
    3. Result: ["à¤¤à¥‡à¤°à¥€", "à¤•à¤¸à¤®", "ğŸ˜€", "ğŸ˜€"]
    """
    # Get words from both texts
    original_words = original_text.split()
    preprocessed_words = preprocessed_text.split()
    
    # Build reverse emoji mapping (Nepali text â†’ emoji)
    reverse_emoji_map = {nepali: emoji_char for emoji_char, nepali in EMOJI_TO_NEPALI.items()}
    
    # Clean and group tokenizer output into words
    word_pieces = []
    current_word = ""
    
    for tok in tokenizer_tokens:
        # Skip special tokens
        if tok in ['<s>', '</s>', '[CLS]', '[SEP]', '<pad>', '[PAD]']:
            continue
        
        if tok.startswith("â–"):
            # New word
            if current_word:
                word_pieces.append(current_word)
            current_word = tok.replace("â–", "")
        else:
            # Continue current word
            current_word += tok.replace("â–", "")
    
    if current_word:
        word_pieces.append(current_word)
    
    # Now map word_pieces (from preprocessed) back to original with emojis
    display_tokens = []
    orig_idx = 0
    
    for word in word_pieces:
        # Check if this word is an emoji translation
        if word in reverse_emoji_map:
            # This is a Nepali emoji translation â†’ use the actual emoji
            display_tokens.append(reverse_emoji_map[word])
        else:
            # Regular word - try to match with original
            matched = False
            
            # Look for matching word in original
            while orig_idx < len(original_words):
                orig_word = original_words[orig_idx]
                
                # Skip emojis in original (they're handled by reverse_emoji_map)
                if any(c in emoji.EMOJI_DATA for c in orig_word):
                    orig_idx += 1
                    continue
                
                # Check if words match (roughly)
                orig_clean = emoji.replace_emoji(orig_word, replace="").strip()
                if orig_clean and (word in orig_clean or orig_clean in word or word == orig_clean):
                    display_tokens.append(orig_word)
                    matched = True
                    orig_idx += 1
                    break
                
                orig_idx += 1
            
            if not matched:
                # Couldn't match - use the word as-is
                display_tokens.append(word)
    
    return display_tokens

# ============================================================================
# FONT HANDLING
# ============================================================================

def apply_nepali_font_to_text(text_obj, nepali_font: Optional[FontProperties] = None):
    """Apply Nepali font to text objects containing Devanagari"""
    if nepali_font is None:
        return
    
    text_content = text_obj.get_text()
    if regex.search(r'\p{Devanagari}', text_content):
        text_obj.set_fontproperties(nepali_font)


# ----------------------------
# Load model
# ----------------------------
def load_model():
    try:
        if os.path.exists(LOCAL_MODEL_PATH):
            tokenizer = AutoTokenizer.from_pretrained(LOCAL_MODEL_PATH)
            model = AutoModelForSequenceClassification.from_pretrained(
                LOCAL_MODEL_PATH
            ).to(DEVICE).eval()
            le_path = os.path.join(LOCAL_MODEL_PATH, "label_encoder.pkl")
            le = joblib.load(le_path) if os.path.exists(le_path) else None
            print("âœ… Model loaded from LOCAL path")
            return model, tokenizer, le
        else:
            tokenizer = AutoTokenizer.from_pretrained(HF_MODEL_ID)
            model = AutoModelForSequenceClassification.from_pretrained(
                HF_MODEL_ID
            ).to(DEVICE).eval()
            try:
                le_file = hf_hub_download(
                    repo_id=HF_MODEL_ID, 
                    filename="label_encoder.pkl", 
                    cache_dir="models/cache"
                )
                le = joblib.load(le_file)
            except:
                from sklearn.preprocessing import LabelEncoder
                le = LabelEncoder()
                le.fit(['NO', 'OO', 'OR', 'OS'])
                print("âš ï¸ Label encoder not found, using default")
            print("âœ… Model loaded from HuggingFace Hub")
            return model, tokenizer, le
    except Exception as e:
        print(f"âŒ Failed to load model: {e}")
        return None, None, None


# ----------------------------
# Captum Explainer with Emoji Support
# ----------------------------
def explain_text_with_captum(
    original_text: str,
    model,
    tokenizer,
    label_encoder,
    preprocessor: HateSpeechPreprocessor,
    target: Optional[int] = None,
    n_steps: int = 50,
    out_dir: str = OUT_DIR,
    nepali_font: Optional[FontProperties] = NEPALI_FONT
):
    """
    Explain text with Captum Integrated Gradients
    
    Args:
        original_text: Original text with emojis
        model: Trained model
        tokenizer: Tokenizer
        label_encoder: Label encoder
        preprocessor: Preprocessing pipeline
        target: Target class (None = predicted class)
        n_steps: Number of IG steps
        out_dir: Output directory
        nepali_font: Font for Nepali text
    
    Returns:
        Dictionary with explanation results
    """
    # Preprocess text
    preprocessed, emoji_features = preprocessor.preprocess(original_text)
    
    if not preprocessed:
        return None
    
    # Tokenize
    encoding = tokenizer(
        preprocessed,
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=MAX_LENGTH
    )
    input_ids = encoding['input_ids'].to(DEVICE)
    attention_mask = encoding['attention_mask'].to(DEVICE)
    
    # Get prediction
    with torch.no_grad():
        out = model(input_ids=input_ids, attention_mask=attention_mask)
        probs = torch.softmax(out.logits, dim=-1)[0].cpu().numpy()
        pred_idx = int(np.argmax(probs))
        pred_label = label_encoder.classes_[pred_idx]
        pred_conf = float(probs[pred_idx])
    
    if target is None:
        target = pred_idx
    
    # Captum Integrated Gradients
    # CRITICAL: Use input_ids, not embeddings directly
    embedding_layer = model.roberta.embeddings.word_embeddings
    
    def forward_func(input_ids_arg, attention_mask_arg):
        """Forward function that takes input_ids"""
        return model(input_ids=input_ids_arg, attention_mask=attention_mask_arg).logits[:, target]
    
    lig = LayerIntegratedGradients(forward_func, embedding_layer)
    
    # Baseline: all pad tokens
    baseline_ids = torch.full_like(input_ids, tokenizer.pad_token_id)
    
    # Calculate attributions using input_ids (not embeddings)
    attributions, delta = lig.attribute(
        input_ids,
        baselines=baseline_ids,
        additional_forward_args=(attention_mask,),
        return_convergence_delta=True,
        n_steps=n_steps
    )
    
    # Sum across embedding dimension
    attributions_sum = attributions.sum(dim=-1).squeeze(0)
    
    # Get tokens
    tokens = tokenizer.convert_ids_to_tokens(
        input_ids[0].cpu().tolist(),
        skip_special_tokens=False
    )
    
    # Create display tokens with emojis preserved
    display_tokens = create_display_tokens_from_subwords(
        original_text,
        preprocessed,
        tokens,
        remove_special=True
    )
    
    # Aggregate word-level attributions
    word_attributions = []
    current_tokens = []
    current_indices = []
    
    for i, tok in enumerate(tokens):
        # Skip special tokens
        if tok in ['<s>', '</s>', '[CLS]', '[SEP]', '<pad>', '[PAD]']:
            continue
        
        if tok.startswith("â–"):
            # New word starts
            if current_indices:
                # Save previous word
                grp_vals = attributions_sum[current_indices].detach().cpu().numpy()
                score = float(np.sum(np.abs(grp_vals)))
                signed_score = float(np.sum(grp_vals))
                word = "".join([tokens[j].replace("â–", "") for j in current_indices])
                word_attributions.append((word, score, signed_score))
            
            current_indices = [i]
            current_tokens = [tok.replace("â–", "")]
        else:
            # Continue current word
            current_indices.append(i)
            current_tokens.append(tok)
    
    # Don't forget last word
    if current_indices:
        grp_vals = attributions_sum[current_indices].detach().cpu().numpy()
        score = float(np.sum(np.abs(grp_vals)))
        signed_score = float(np.sum(grp_vals))
        word = "".join([tokens[j].replace("â–", "") for j in current_indices])
        word_attributions.append((word, score, signed_score))
    
    # Align attributions with display tokens (with emojis)
    if len(display_tokens) == len(word_attributions):
        # Direct alignment - replace words with display tokens
        aligned_attributions = [
            (display_tok, score, signed_score)
            for display_tok, (_, score, signed_score) in zip(display_tokens, word_attributions)
        ]
    else:
        # Use word_attributions as-is if mismatch
        aligned_attributions = word_attributions
    
    # Visualize
    visualize_captum_results(
        aligned_attributions,
        pred_label,
        pred_conf,
        original_text,
        preprocessed,
        emoji_features,
        out_dir,
        nepali_font
    )
    
    return {
        "original_text": original_text,
        "preprocessed": preprocessed,
        "emoji_features": emoji_features,
        "pred_label": pred_label,
        "pred_confidence": pred_conf,
        "word_attributions": aligned_attributions,
        "convergence_delta": float(delta.sum().cpu().numpy())
    }


def visualize_captum_results(
    word_attributions: List[Tuple[str, float, float]],
    pred_label: str,
    pred_conf: float,
    original_text: str,
    preprocessed_text: str,
    emoji_features: Dict[str, int],
    out_dir: str,
    nepali_font: Optional[FontProperties]
):
    """Create visualizations for Captum results"""
    
    scores_plot = [s for _, s, _ in word_attributions]
    words_plot = [w for w, _, _ in word_attributions]
    max_score = max(scores_plot) if scores_plot else 1.0
    
    # ===== BAR CHART =====
    fig, ax = plt.subplots(figsize=(max(8, 0.6 * len(words_plot)), 5))
    colors = ['green' if ss > 0 else 'red' 
              for _, _, ss in word_attributions]
    
    ax.bar(range(len(words_plot)), scores_plot, tick_label=words_plot, color=colors, alpha=0.7)
    ax.set_ylabel("Attribution (sum abs)", fontsize=12)
    ax.set_title(
        f"Integrated Gradients â†’ Pred: {pred_label} ({pred_conf:.2%})",
        fontsize=14,
        fontweight='bold'
    )
    
    # Apply Nepali font to Devanagari text in tick labels (emojis render naturally)
    if nepali_font:
        for lbl in ax.get_xticklabels():
            text_content = lbl.get_text()
            # Only apply font to Devanagari, not emojis
            if regex.search(r"\p{Devanagari}", text_content):
                lbl.set_fontproperties(nepali_font)
            lbl.set_fontsize(11)
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    
    barpath = os.path.join(
        out_dir,
        f"ig_bar_{pred_label}_{abs(hash(original_text)) % 10**8}.png"
    )
    fig.savefig(barpath, dpi=300, bbox_inches='tight')
    
    # SHOW PLOT INLINE
    plt.show()
    
    plt.close(fig)
    print(f"   ğŸ“Š Bar chart: {barpath}")
    
    # ===== TEXT HEATMAP =====
    cmap = cm.get_cmap("RdYlGn")
    
    fig2, ax2 = plt.subplots(figsize=(max(10, 0.6 * len(words_plot)), 3))
    ax2.axis('off')
    
    x, y = 0.01, 0.6
    text_objs = []
    
    for word, score, signed_score in word_attributions:
        # Normalize for color
        intensity = min(score / max_score, 1.0) if max_score > 0 else 0.0
        
        # Color based on signed score
        if signed_score > 0:
            color = cmap(0.5 + intensity * 0.5)  # Green side
        else:
            color = cmap(0.5 - intensity * 0.5)  # Red side
        
        txt = ax2.text(
            x, y, f" {word} ",
            fontsize=13,
            bbox=dict(
                facecolor=mcolors.to_hex(color),
                alpha=0.8,
                boxstyle="round,pad=0.3",
                edgecolor='gray'
            )
        )
        
        # Apply Nepali font only to Devanagari text (not emojis)
        if nepali_font and regex.search(r"\p{Devanagari}", word):
            txt.set_fontproperties(nepali_font)
        
        text_objs.append(txt)
        
        # Update position - emojis take less horizontal space
        char_width = 0.025 if any(c in emoji.EMOJI_DATA for c in word) else 0.04
        x += char_width * len(word) + 0.01
        
        if x > 0.92:
            x = 0.01
            y -= 0.35
    
    # Title
    ax2.text(
        0.5, 0.95,
        f"Token Attributions (Predicted: {pred_label})",
        ha='center',
        va='top',
        fontsize=14,
        fontweight='bold',
        transform=ax2.transAxes
    )
    
    plt.tight_layout()
    
    heatpath = os.path.join(
        out_dir,
        f"ig_heatmap_{pred_label}_{abs(hash(original_text)) % 10**8}.png"
    )
    fig2.savefig(heatpath, dpi=300, bbox_inches='tight')
    
    # SHOW PLOT INLINE
    plt.show()
    
    plt.close(fig2)
    print(f"   ğŸ¨ Heatmap: {heatpath}")


# ----------------------------
# Main execution
# ----------------------------
if __name__ == "__main__":
    # Load model
    model, tokenizer, le = load_model()
    if model is None:
        raise RuntimeError("Model not loaded")
    
    # Initialize preprocessor
    preprocessor = HateSpeechPreprocessor(
        model_type="xlmr",
        translate_english=False,
        cache_size=2000
    )
    
    # Load test data
    df = pd.read_json(TEST_FILE)
    pd.set_option("display.max_colwidth", None)
    
    # ----------------------------
    # CONFIG: samples per class
    # ----------------------------
    SAMPLES_PER_CLASS = 3
    LABEL_COL = "Label_Multiclass"
    TEXT_COL = "Comment"
    
    # ----------------------------
    # Sample per class with emoji priority
    # ----------------------------
    sampled_rows = []
    
    for label, group in df.groupby(LABEL_COL):
        # Prioritize rows with emojis
        has_emoji = group[group[TEXT_COL].apply(
            lambda x: any(c in emoji.EMOJI_DATA for c in str(x))
        )]
        no_emoji = group.drop(has_emoji.index)
        
        selected = []
        
        # Pick one with emoji if available
        if len(has_emoji) > 0:
            selected.append(has_emoji.sample(n=1, random_state=SEED))
        
        # Fill remaining samples
        remaining_needed = SAMPLES_PER_CLASS - len(selected)
        if remaining_needed > 0:
            pool = no_emoji if len(no_emoji) >= remaining_needed else group
            selected.append(pool.sample(
                n=min(len(pool), remaining_needed),
                random_state=SEED
            ))
        
        if selected:
            sampled_rows.append(pd.concat(selected))
    
    sampled_df = pd.concat(sampled_rows) if sampled_rows else df.head(5)
    
    print("âœ… Samples per class:")
    print(sampled_df[LABEL_COL].value_counts())
    print(f"\n{'='*80}")
    
    # ----------------------------
    # Run Captum explanations
    # ----------------------------
    for idx, row in tqdm(sampled_df.iterrows(), total=len(sampled_df), desc="Explaining"):
        text = str(row[TEXT_COL]).strip()
        true_label = row[LABEL_COL]
        
        if not text:
            continue
        
        print(f"\n{'='*80}")
        print(f"Sample {idx + 1}")
        print(f"{'='*80}")
        print(f"ğŸ“ Original: {text}")
        print(f"ğŸ¯ True Label: {true_label}")
        
        try:
            res = explain_text_with_captum(
                original_text=text,
                model=model,
                tokenizer=tokenizer,
                label_encoder=le,
                preprocessor=preprocessor,
                n_steps=50,
                out_dir=OUT_DIR,
                nepali_font=NEPALI_FONT
            )
            
            if res:
                print(f"ğŸ”„ Preprocessed: {res['preprocessed']}")
                print(f"\nğŸ˜Š Emoji Features:")
                for key, val in res['emoji_features'].items():
                    if val > 0 or 'count' in key:
                        print(f"   {key}: {val}")
                print(f"\nğŸ¤– Predicted: {res['pred_label']} ({res['pred_confidence']:.4f})")
                print(f"ğŸ“ˆ Convergence Delta: {res['convergence_delta']:.6f}")
                print(f"âœ… Explanation saved")
        
        except Exception as e:
            print(f"âš ï¸ Failed for sample {idx}: {e}")
            import traceback
            traceback.print_exc()
    
    print(f"\n{'='*80}")
    print(f"âœ… All Captum explanations completed!")
    print(f"ğŸ“ Outputs saved to: {OUT_DIR}")
```
