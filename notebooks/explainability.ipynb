```python
# =============================
# LIME & SHAP Explainer for Kaggle (Nepali Font + Emoji Friendly)
# Updated: Integrated with new preprocessing pipeline + Emoji visualization
# =============================

import os
import numpy as np
import pandas as pd
import torch
import re
import emoji
import regex
import warnings
warnings.filterwarnings("ignore")
from typing import Optional, Tuple, Dict, List
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties, fontManager

from transformers import AutoTokenizer, AutoModelForSequenceClassification
from huggingface_hub import hf_hub_download
import joblib

# ============================ FONT CONFIG ==================================
FONT_PATH = '/kaggle/input/fontss/Kalimati.ttf'  # Kaggle uploaded font

def load_nepali_font(font_path):
    if not os.path.exists(font_path):
        print(f"âš ï¸ Nepali font not found: {font_path}")
        return None
    fontManager.addfont(font_path)
    fp = FontProperties(fname=font_path)
    print(f"âœ“ Loaded Nepali font: {fp.get_name()}")
    return fp

NEPALI_FONT = load_nepali_font(FONT_PATH)
plt.rcParams["axes.unicode_minus"] = False

OUT_DIR = '/kaggle/working/explanations'
os.makedirs(OUT_DIR, exist_ok=True)

# ============================ LIME / SHAP ==================================
try:
    from lime.lime_text import LimeTextExplainer
    LIME_AVAILABLE = True
except ImportError:
    LIME_AVAILABLE = False
    print("âš ï¸ LIME not installed")

try:
    import shap
    SHAP_AVAILABLE = True
except ImportError:
    SHAP_AVAILABLE = False
    print("âš ï¸ SHAP not installed")

# ============================ PREPROCESSING PIPELINE ========================
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
from deep_translator import GoogleTranslator
from functools import lru_cache
import logging

logging.basicConfig(level=logging.WARNING)
logger = logging.getLogger(__name__)

try:
    from indic_transliteration import sanscript
    from indic_transliteration.sanscript import transliterate
    TRANSLITERATION_AVAILABLE = True
except ImportError:
    TRANSLITERATION_AVAILABLE = False
    logger.warning("indic_transliteration not available. Transliteration disabled.")


# ============================================================================
# ENHANCED EMOJI MAPPING (Semantic & Context-Aware)
# ============================================================================

EMOJI_TO_NEPALI = {
    # ===== Laughter & Joy (Positive) =====
    'ğŸ˜‚': 'à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ¤£': 'à¤ à¥‚à¤²à¥‹ à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ˜€': 'à¤–à¥à¤¶à¥€', 'ğŸ˜': 'à¤–à¥à¤¶à¥€', 'ğŸ˜ƒ': 'à¤–à¥à¤¶à¥€',
    'ğŸ˜„': 'à¤–à¥à¤¶à¥€', 'ğŸ˜…': 'à¤¨à¤°à¥à¤­à¤¸ à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ˜†': 'à¤¹à¤¾à¤à¤¸à¥‹', 'ğŸ˜Š': 'à¤®à¥à¤¸à¥à¤•à¤¾à¤¨', 'â˜ºï¸': 'à¤®à¥à¤¸à¥à¤•à¤¾à¤¨',
    'ğŸ˜‰': 'à¤†à¤à¤–à¤¾ à¤à¤¿à¤®à¥à¤•à¤¾à¤‰à¤¨à¥‡', 'ğŸ™‚': 'à¤®à¥à¤¸à¥à¤•à¤¾à¤¨', 'ğŸ™ƒ': 'à¤‰à¤²à¥à¤Ÿà¥‹ à¤®à¥à¤¸à¥à¤•à¤¾à¤¨', 'ğŸ˜Œ': 'à¤¶à¤¾à¤¨à¥à¤¤',
    'ğŸ˜': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ¥°': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ˜˜': 'à¤šà¥à¤®à¥à¤¬à¤¨', 'ğŸ˜—': 'à¤šà¥à¤®à¥à¤¬à¤¨', 'ğŸ˜™': 'à¤šà¥à¤®à¥à¤¬à¤¨', 'ğŸ˜š': 'à¤šà¥à¤®à¥à¤¬à¤¨',
    'ğŸ¤—': 'à¤…à¤à¤—à¤¾à¤²à¥‹', 'ğŸ¤©': 'à¤šà¤•à¤¿à¤¤', 'ğŸ¥³': 'à¤‰à¤¤à¥à¤¸à¤µ', 'ğŸ¤¤': 'à¤²à¤¾à¤²à¤¸à¤¾',
    
    # ===== Mocking & Sarcasm (Hate Indicator) =====
    'ğŸ˜': 'à¤µà¥à¤¯à¤‚à¤—à¥à¤¯', 'ğŸ˜œ': 'à¤œà¤¿à¤¬à¥à¤°à¥‹ à¤¦à¥‡à¤–à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜': 'à¤œà¤¿à¤¬à¥à¤°à¥‹ à¤¦à¥‡à¤–à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜›': 'à¤œà¤¿à¤¬à¥à¤°à¥‹',
    'ğŸ™„': 'à¤†à¤à¤–à¤¾ à¤˜à¥à¤®à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜¤': 'à¤¨à¤¿à¤°à¤¾à¤¶', 'ğŸ˜‘': 'à¤…à¤­à¤¿à¤µà¥à¤¯à¤•à¥à¤¤à¤¿à¤¹à¥€à¤¨', 'ğŸ˜': 'à¤¤à¤Ÿà¤¸à¥à¤¥',
    'ğŸ˜¬': 'à¤¤à¤¨à¤¾à¤µ', 'ğŸ¤¨': 'à¤¶à¤‚à¤•à¤¾à¤¸à¥à¤ªà¤¦', 'ğŸ¤«': 'à¤šà¥à¤ªà¤šà¤¾à¤ª',
    'ğŸ¤­': 'à¤¹à¤¾à¤¤ à¤›à¥‹à¤ªà¥à¤¨à¥‡', 'ğŸ¤¥': 'à¤à¥‚à¤ ', 'ğŸ˜¶': 'à¤®à¥Œà¤¨',
    
    # ===== Anger & Hate (CRITICAL for Detection) =====
    'ğŸ˜ ': 'à¤°à¤¿à¤¸', 'ğŸ˜¡': 'à¤ à¥‚à¤²à¥‹ à¤°à¤¿à¤¸', 'ğŸ¤¬': 'à¤—à¤¾à¤²à¥€', 'ğŸ˜ˆ': 'à¤–à¤°à¤¾à¤¬', 'ğŸ‘¿': 'à¤–à¤°à¤¾à¤¬',
    'ğŸ’¢': 'à¤•à¥à¤°à¥‹à¤§', 'ğŸ”ª': 'à¤¹à¤¿à¤‚à¤¸à¤¾', 'ğŸ’£': 'à¤¹à¤¿à¤‚à¤¸à¤¾', 'ğŸ—¡ï¸': 'à¤¤à¤°à¤µà¤¾à¤°', 'âš”ï¸': 'à¤¯à¥à¤¦à¥à¤§',
    'ğŸ’¥': 'à¤µà¤¿à¤¸à¥à¤«à¥‹à¤Ÿ', 'ğŸ”«': 'à¤¬à¤¨à¥à¤¦à¥à¤•', 'ğŸ§¨': 'à¤µà¤¿à¤¸à¥à¤«à¥‹à¤Ÿà¤•', 'â˜ ï¸': 'à¤®à¥ƒà¤¤à¥à¤¯à¥', 'ğŸ’€': 'à¤–à¥‹à¤ªà¤¡à¥€',
    'ğŸ‘¹': 'à¤°à¤¾à¤•à¥à¤·à¤¸', 'ğŸ‘º': 'à¤¦à¤¾à¤¨à¤µ', 'ğŸ¤¡': 'à¤œà¥‹à¤•à¤°', 'ğŸ–¤': 'à¤•à¤¾à¤²à¥‹ à¤®à¤¨',
    'ğŸ˜¾': 'à¤°à¤¿à¤¸à¤¾à¤à¤•à¥‹', 'ğŸ‘Š': 'à¤®à¥à¤•à¥à¤•à¤¾', 'âœŠ': 'à¤®à¥à¤•à¥à¤•à¤¾',
    
    # ===== Offensive Gestures (Strong Hate Signal) =====
    'ğŸ–•': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ‘': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ»': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¼': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•',
    'ğŸ‘ğŸ½': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¾': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¿': 'à¤¨à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•',
    'ğŸ–•ğŸ»': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ¼': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ½': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ¾': 'à¤…à¤ªà¤®à¤¾à¤¨', 'ğŸ–•ğŸ¿': 'à¤…à¤ªà¤®à¤¾à¤¨',
    
    # ===== Sadness & Crying =====
    'ğŸ˜­': 'à¤°à¥à¤µà¤¾à¤‡', 'ğŸ˜¢': 'à¤°à¥à¤µà¤¾à¤‡', 'ğŸ˜¿': 'à¤°à¥à¤µà¤¾à¤‡', 'ğŸ˜”': 'à¤‰à¤¦à¤¾à¤¸', 'ğŸ˜': 'à¤‰à¤¦à¤¾à¤¸',
    'ğŸ˜’': 'à¤‰à¤¦à¤¾à¤¸', 'ğŸ˜“': 'à¤šà¤¿à¤¨à¥à¤¤à¤¿à¤¤', 'ğŸ˜Ÿ': 'à¤šà¤¿à¤¨à¥à¤¤à¤¿à¤¤', 'ğŸ˜•': 'à¤…à¤²à¤®à¤²à¤¿à¤à¤•à¥‹',
    'ğŸ™': 'à¤¤à¤²à¥à¤²à¥‹ à¤®à¥à¤–', 'â˜¹ï¸': 'à¤¦à¥à¤ƒà¤–à¥€', 'ğŸ˜©': 'à¤¥à¤•à¤¿à¤¤', 'ğŸ˜«': 'à¤¥à¤•à¤¿à¤¤',
    'ğŸ˜–': 'à¤­à¥à¤°à¤®à¤¿à¤¤', 'ğŸ˜£': 'à¤…à¤¡à¤¿à¤—', 'ğŸ˜¥': 'à¤¨à¤¿à¤°à¤¾à¤¶', 'ğŸ¥º': 'à¤¬à¤¿à¤¨à¥à¤¤à¥€',
    
    # ===== Fear & Shock =====
    'ğŸ˜¨': 'à¤¡à¤°', 'ğŸ˜°': 'à¤šà¤¿à¤¨à¥à¤¤à¤¿à¤¤ à¤ªà¤¸à¤¿à¤¨à¤¾', 'ğŸ˜±': 'à¤šà¤¿à¤šà¥à¤¯à¤¾à¤‰à¤¨à¥‡', 'ğŸ˜³': 'à¤²à¤œà¤¾à¤‰à¤¨à¥‡',
    'ğŸ¤¯': 'à¤®à¤¨ à¤‰à¤¡à¥‡à¤•à¥‹', 'ğŸ˜µ': 'à¤šà¤•à¥à¤•à¤°', 'ğŸ˜²': 'à¤šà¤•à¤¿à¤¤', 'ğŸ˜¯': 'à¤›à¤•à¥à¤•',
    
    # ===== Disgust & Contempt =====
    'ğŸ¤¢': 'à¤¬à¤¾à¤¨à¥à¤¤à¤¾', 'ğŸ¤®': 'à¤¬à¤¾à¤¨à¥à¤¤à¤¾', 'ğŸ¤§': 'à¤¹à¤¾à¤šà¥à¤›à¥à¤¯à¥‚à¤', 'ğŸ˜·': 'à¤¬à¤¿à¤°à¤¾à¤®à¥€',
    'ğŸ¤’': 'à¤œà¥à¤µà¤°à¥‹', 'ğŸ¤•': 'à¤˜à¤¾à¤‡à¤¤à¥‡', 'ğŸ¥´': 'à¤®à¤¾à¤¤à¥à¤¤à¤¿à¤à¤•à¥‹', 'ğŸ˜ª': 'à¤¨à¤¿à¤¦à¥à¤°à¤¾',
    
    # ===== Positive Gestures & Symbols =====
    'ğŸ‘': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ»': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¼': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 
    'ğŸ‘ğŸ½': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¾': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•', 'ğŸ‘ğŸ¿': 'à¤¸à¤•à¤¾à¤°à¤¾à¤¤à¥à¤®à¤•',
    'ğŸ‘': 'à¤¤à¤¾à¤²à¤¿', 'ğŸ™Œ': 'à¤‰à¤¤à¥à¤¸à¤µ', 'ğŸ‘Œ': 'à¤ à¥€à¤• à¤›', 'ğŸ¤': 'à¤¹à¤¾à¤¤ à¤®à¤¿à¤²à¤¾à¤‰à¤¨à¥',
    'ğŸ™': 'à¤¨à¤®à¤¸à¥à¤•à¤¾à¤°', 'ğŸ¤²': 'à¤ªà¥à¤°à¤¾à¤°à¥à¤¥à¤¨à¤¾', 'ğŸ’ª': 'à¤¶à¤•à¥à¤¤à¤¿', 'âœŒï¸': 'à¤¶à¤¾à¤¨à¥à¤¤à¤¿',
    
    # ===== Hearts & Love =====
    'â¤ï¸': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ§¡': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’›': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’š': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’™': 'à¤®à¤¾à¤¯à¤¾',
    'ğŸ’œ': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ¤': 'à¤¸à¥‡à¤¤à¥‹ à¤®à¤¨', 'ğŸ¤': 'à¤–à¥ˆà¤°à¥‹ à¤®à¤¨',
    'â£ï¸': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’•': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’“': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’—': 'à¤®à¤¾à¤¯à¤¾',
    'ğŸ’–': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’˜': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’': 'à¤®à¤¾à¤¯à¤¾', 'ğŸ’”': 'à¤Ÿà¥à¤Ÿà¥‡à¤•à¥‹ à¤®à¤¨',
    
    # ===== Symbols & Objects =====
    'ğŸ”¥': 'à¤†à¤—à¥‹', 'ğŸ’¯': 'à¤ªà¥‚à¤°à¥à¤£', 'ğŸ’¨': 'à¤¹à¤¾à¤µà¤¾', 'ğŸ’«': 'à¤šà¤®à¤•',
    'â­': 'à¤¤à¤¾à¤°à¤¾', 'âœ¨': 'à¤šà¤®à¤•', 'ğŸŒŸ': 'à¤šà¤®à¥à¤•à¤¿à¤²à¥‹ à¤¤à¤¾à¤°à¤¾',
    'ğŸš«': 'à¤¨à¤¿à¤·à¥‡à¤§', 'â›”': 'à¤ªà¥à¤°à¤µà¥‡à¤¶ à¤¨à¤¿à¤·à¥‡à¤§', 'âŒ': 'à¤°à¤¦à¥à¤¦', 'â': 'à¤—à¤²à¤¤',
    
    # ===== People & Relationships =====
    'ğŸ‘«': 'à¤œà¥‹à¤¡à¥€', 'ğŸ‘¬': 'à¤ªà¥à¤°à¥à¤· à¤œà¥‹à¤¡à¥€', 'ğŸ‘­': 'à¤®à¤¹à¤¿à¤²à¤¾ à¤œà¥‹à¤¡à¥€', 'ğŸ‘¨â€ğŸ‘©â€ğŸ‘§â€ğŸ‘¦': 'à¤ªà¤°à¤¿à¤µà¤¾à¤°',
    'ğŸ‘¶': 'à¤¬à¤šà¥à¤šà¤¾', 'ğŸ‘¦': 'à¤•à¥‡à¤Ÿà¤¾', 'ğŸ‘§': 'à¤•à¥‡à¤Ÿà¥€', 'ğŸ‘¨': 'à¤ªà¥à¤°à¥à¤·', 'ğŸ‘©': 'à¤®à¤¹à¤¿à¤²à¤¾',
    'ğŸ‘´': 'à¤¬à¥‚à¤¢à¥‹', 'ğŸ‘µ': 'à¤¬à¥‚à¤¢à¥€', 'ğŸ§’': 'à¤¬à¤¾à¤²à¤•', 'ğŸ‘±': 'à¤—à¥‹à¤°à¥‹', 'ğŸ§”': 'à¤¦à¤¾à¤¹à¥à¤°à¥€',
    
    # ===== Country & Culture =====
    'ğŸ‡³ğŸ‡µ': 'à¤¨à¥‡à¤ªà¤¾à¤²', 'ğŸ‡®ğŸ‡³': 'à¤­à¤¾à¤°à¤¤', 'ğŸ‡µğŸ‡°': 'à¤ªà¤¾à¤•à¤¿à¤¸à¥à¤¤à¤¾à¤¨', 'ğŸ‡§ğŸ‡©': 'à¤¬à¤‚à¤—à¤²à¤¾à¤¦à¥‡à¤¶',
    'ğŸ‡¨ğŸ‡³': 'à¤šà¥€à¤¨', 'ğŸ‡ºğŸ‡¸': 'à¤…à¤®à¥‡à¤°à¤¿à¤•à¤¾', 'ğŸ´': 'à¤à¤£à¥à¤¡à¤¾',
    
    # ===== Animals (sometimes used in hate) =====
    'ğŸ•': 'à¤•à¥à¤•à¥à¤°', 'ğŸ–': 'à¤¸à¥à¤à¤—à¥à¤°', 'ğŸ€': 'à¤®à¥à¤¸à¤¾', 'ğŸ': 'à¤¸à¤°à¥à¤ª', 'ğŸ¦‚': 'à¤¬à¤¿à¤šà¥à¤›à¥€',
    'ğŸ’': 'à¤¬à¤¾à¤à¤¦à¤°', 'ğŸµ': 'à¤¬à¤¾à¤à¤¦à¤° à¤…à¤¨à¥à¤¹à¤¾à¤°', 'ğŸ¦': 'à¤—à¥‹à¤°à¤¿à¤²à¥à¤²à¤¾', 'ğŸ—': 'à¤œà¤™à¥à¤—à¤²à¥€ à¤¸à¥à¤à¤—à¥à¤°',
    
    # ===== Other Common =====
    'ğŸ¤”': 'à¤¸à¥‹à¤š', 'ğŸ§': 'à¤…à¤¨à¥à¤¸à¤¨à¥à¤§à¤¾à¤¨', 'ğŸ˜´': 'à¤¸à¥à¤¤à¥à¤¨à¥‡', 'ğŸ’©': 'à¤®à¤²',
    'ğŸ‘»': 'à¤­à¥‚à¤¤', 'ğŸ¤–': 'à¤°à¥‹à¤¬à¥‹à¤Ÿ', 'ğŸ‘½': 'à¤µà¤¿à¤¦à¥‡à¤¶à¥€', 'ğŸ­': 'à¤®à¥à¤–à¥Œà¤Ÿà¤¾',
}

# Emoji categories for INTELLIGENT FEATURE EXTRACTION
HATE_RELATED_EMOJIS = {
    'ğŸ˜ ', 'ğŸ˜¡', 'ğŸ¤¬', 'ğŸ˜ˆ', 'ğŸ‘¿', 'ğŸ’¢', 'ğŸ‘Š', 'âœŠ',
    'ğŸ”ª', 'ğŸ’£', 'ğŸ—¡ï¸', 'âš”ï¸', 'ğŸ’¥', 'ğŸ”«', 'ğŸ§¨', 'â˜ ï¸', 'ğŸ’€',
    'ğŸ–•', 'ğŸ–•ğŸ»', 'ğŸ–•ğŸ¼', 'ğŸ–•ğŸ½', 'ğŸ–•ğŸ¾', 'ğŸ–•ğŸ¿',
    'ğŸ‘', 'ğŸ‘ğŸ»', 'ğŸ‘ğŸ¼', 'ğŸ‘ğŸ½', 'ğŸ‘ğŸ¾', 'ğŸ‘ğŸ¿',
    'ğŸ‘¹', 'ğŸ‘º', 'ğŸ¤¡', 'ğŸ–¤', 'ğŸ’”',
    'ğŸ•', 'ğŸ–', 'ğŸ€', 'ğŸ', 'ğŸ¦‚', 'ğŸ’', 'ğŸµ', 'ğŸ¦', 'ğŸ—',
    'ğŸ’©', 'ğŸ˜¾',
}

MOCKERY_EMOJIS = {
    'ğŸ˜', 'ğŸ˜œ', 'ğŸ˜', 'ğŸ˜›', 'ğŸ™„', 'ğŸ˜¤', 'ğŸ™ƒ',
    'ğŸ˜‘', 'ğŸ˜', 'ğŸ˜¬', 'ğŸ¤¨', 'ğŸ¤«', 'ğŸ¤­', 'ğŸ¤¥',
    'ğŸ¤¡', 'ğŸ‘»', 'ğŸ­',
}

POSITIVE_EMOJIS = {
    'ğŸ˜Š', 'ğŸ˜€', 'ğŸ˜', 'ğŸ˜ƒ', 'ğŸ˜„', 'â˜ºï¸', 'ğŸ™‚', 'ğŸ˜Œ', 'ğŸ¥°', 'ğŸ˜',
    'â¤ï¸', 'ğŸ§¡', 'ğŸ’›', 'ğŸ’š', 'ğŸ’™', 'ğŸ’œ', 'ğŸ¤', 'ğŸ¤',
    'ğŸ’•', 'ğŸ’', 'ğŸ’“', 'ğŸ’—', 'ğŸ’–', 'ğŸ’˜', 'ğŸ’', 'â£ï¸',
    'ğŸ‘', 'ğŸ‘ğŸ»', 'ğŸ‘ğŸ¼', 'ğŸ‘ğŸ½', 'ğŸ‘ğŸ¾', 'ğŸ‘ğŸ¿',
    'ğŸ™', 'ğŸ‘', 'ğŸ™Œ', 'ğŸ‘Œ', 'ğŸ¤', 'âœŒï¸',
    'ğŸ¥³', 'ğŸ‰', 'ğŸŠ', 'â­', 'âœ¨', 'ğŸŒŸ',
}

SADNESS_EMOJIS = {
    'ğŸ˜­', 'ğŸ˜¢', 'ğŸ˜¿', 'ğŸ˜”', 'ğŸ˜', 'ğŸ˜’', 'ğŸ˜“', 'ğŸ˜Ÿ', 'ğŸ˜•',
    'ğŸ™', 'â˜¹ï¸', 'ğŸ˜©', 'ğŸ˜«', 'ğŸ˜–', 'ğŸ˜£', 'ğŸ˜¥', 'ğŸ¥º',
}

FEAR_EMOJIS = {
    'ğŸ˜¨', 'ğŸ˜°', 'ğŸ˜±', 'ğŸ˜³', 'ğŸ¤¯', 'ğŸ˜µ', 'ğŸ˜²', 'ğŸ˜¯',
}

DISGUST_EMOJIS = {
    'ğŸ¤¢', 'ğŸ¤®', 'ğŸ¤§', 'ğŸ˜·', 'ğŸ¤’', 'ğŸ¤•', 'ğŸ¥´',
}


# ============================================================================
# EMOJI INTELLIGENCE - FEATURE EXTRACTION
# ============================================================================

def extract_emoji_features(text: str) -> Dict[str, int]:
    """Extract comprehensive emoji-based semantic features"""
    emojis_found = [c for c in text if c in emoji.EMOJI_DATA]
    
    hate_count = sum(1 for e in emojis_found if e in HATE_RELATED_EMOJIS)
    mockery_count = sum(1 for e in emojis_found if e in MOCKERY_EMOJIS)
    positive_count = sum(1 for e in emojis_found if e in POSITIVE_EMOJIS)
    sadness_count = sum(1 for e in emojis_found if e in SADNESS_EMOJIS)
    fear_count = sum(1 for e in emojis_found if e in FEAR_EMOJIS)
    disgust_count = sum(1 for e in emojis_found if e in DISGUST_EMOJIS)
    
    return {
        'has_hate_emoji': 1 if hate_count > 0 else 0,
        'has_mockery_emoji': 1 if mockery_count > 0 else 0,
        'has_positive_emoji': 1 if positive_count > 0 else 0,
        'has_sadness_emoji': 1 if sadness_count > 0 else 0,
        'has_fear_emoji': 1 if fear_count > 0 else 0,
        'has_disgust_emoji': 1 if disgust_count > 0 else 0,
        'hate_emoji_count': hate_count,
        'mockery_emoji_count': mockery_count,
        'positive_emoji_count': positive_count,
        'sadness_emoji_count': sadness_count,
        'fear_emoji_count': fear_count,
        'disgust_emoji_count': disgust_count,
        'total_emoji_count': len(emojis_found),
        'hate_to_positive_ratio': hate_count / max(positive_count, 1),
        'has_mixed_sentiment': 1 if (hate_count > 0 and positive_count > 0) else 0,
    }


# ============================================================================
# SCRIPT DETECTION (IGNORES EMOJIS)
# ============================================================================

from typing import Literal
ScriptType = Literal["devanagari", "romanized_nepali", "english", "mixed", "other"]


def remove_emojis_for_detection(text: str) -> str:
    """Remove emojis temporarily for script detection"""
    return emoji.replace_emoji(text, replace="")


def detect_script_type(text: str) -> Tuple[ScriptType, dict]:
    """Detect the dominant script type IGNORING emojis"""
    if not text or not text.strip():
        return "other", {"confidence": 0.0, "reason": "empty_text"}
    
    text_no_emoji = remove_emojis_for_detection(text)
    
    if not text_no_emoji.strip():
        return "other", {"confidence": 0.5, "reason": "emoji_only"}
    
    letters = regex.findall(r"\p{L}", text_no_emoji)
    letter_count = len(letters)
    
    if letter_count == 0:
        return "other", {"confidence": 0.0, "reason": "no_letters"}
    
    devanagari_chars = regex.findall(r"\p{Devanagari}", text_no_emoji)
    dev_count = len(devanagari_chars)
    dev_ratio = dev_count / letter_count
    
    latin_chars = regex.findall(r"[a-zA-Z]", text_no_emoji)
    latin_count = len(latin_chars)
    latin_ratio = latin_count / letter_count
    
    romanized_nepali_patterns = [
        r'\b[xX]u\b', r'\b[xX]um?\b', r'\bhajur\b', r'\bdai\b', r'\bbhai\b', r'\bdidi\b',
        r'\bbahini\b', r'\bsanghai\b', r'\bsunu\b', r'\bhera\b', r'\bsun\b',
        r'\bko\b', r'\bki\b', r'\bka\b', r'\bho\b', r'\btyo\b', r'\byo\b', r'\bta\b',
        r'\bma\b', r'\bma?i\b', r'\bla[ie]?\b', r'\bnai?\b', r'\bpani\b', r'\bni\b',
        r'\bhun[ae]\b', r'\bhunchha\b', r'\bhunuhunchha\b', r'\bgar\w+\b', r'\bgarna\b',
        r'\bx[ao]\b', r'\bxa\b', r'\bxan\b', r'\bxaina\b', r'\bxu\b',
        r'\bchain\b', r'\bchaina\b', r'\bthiy[oe]\b', r'\bhola\b', r'\bhos\b',
        r'\bbhan\w*\b', r'\bbol\w*\b', r'\bher\w*\b',
    ]
    
    romanized_indicators = sum(1 for pattern in romanized_nepali_patterns 
                               if re.search(pattern, text_no_emoji, re.IGNORECASE))
    
    romanized_score = 0.0
    if latin_ratio > 0.5 and dev_ratio < 0.3:
        if romanized_indicators > 0:
            romanized_score = min(0.5 + (romanized_indicators * 0.15), 0.95)
        else:
            romanized_patterns = re.findall(r'\b\w*[aeiou](?:h)?\b', text_no_emoji.lower())
            if any(word.endswith(('xu', 'ro', 'no', 'lo', 'ko', 'ho')) 
                   for word in romanized_patterns):
                romanized_score = 0.4
            else:
                romanized_score = 0.3
    
    english_indicators = [
        'the', 'a', 'an', 'this', 'that', 'is', 'am', 'are', 'was', 'were',
        'i', 'you', 'he', 'she', 'it', 'we', 'they', 'what', 'when', 'where',
        'and', 'or', 'but', 'not', 'in', 'on', 'at', 'to', 'for', 'of',
    ]
    english_words = [w.lower() for w in re.findall(r'\b\w+\b', text_no_emoji)]
    english_count = sum(1 for w in english_words if w in english_indicators)
    english_ratio = english_count / len(english_words) if english_words else 0
    
    details = {
        "devanagari_count": dev_count,
        "devanagari_ratio": dev_ratio,
        "latin_count": latin_count,
        "latin_ratio": latin_ratio,
        "romanized_indicators": romanized_indicators,
        "english_ratio": english_ratio,
        "letter_count": letter_count
    }
    
    if dev_ratio >= 0.8:
        return "devanagari", {**details, "confidence": dev_ratio, "reason": "dominant_devanagari"}
    elif dev_ratio >= 0.4:
        return "mixed", {**details, "confidence": 0.7, "reason": "mixed_with_devanagari"}
    elif romanized_score > 0.5 and dev_ratio < 0.2:
        return "romanized_nepali", {**details, "confidence": romanized_score, "reason": "romanized_nepali_detected"}
    elif english_ratio > 0.2 and romanized_score < 0.4:
        return "english", {**details, "confidence": min(english_ratio + 0.3, 0.9), "reason": "english_detected"}
    elif latin_ratio > 0.5 and romanized_score > 0.3:
        return "romanized_nepali", {**details, "confidence": romanized_score, "reason": "likely_romanized_nepali"}
    elif latin_ratio > 0.8:
        if english_ratio > 0.1:
            return "english", {**details, "confidence": 0.6, "reason": "likely_english"}
        else:
            return "romanized_nepali", {**details, "confidence": 0.5, "reason": "ambiguous_latin_script"}
    else:
        return "other", {**details, "confidence": 0.3, "reason": "insufficient_indicators"}


# ============================================================================
# CACHED TRANSLATOR
# ============================================================================

class CachedNepaliTranslator:
    """Production-ready translator with caching"""
    
    def __init__(self, cache_size: int = 2000):
        self.translator = GoogleTranslator(source='en', target='ne')
        self.cache_size = cache_size
        self._translate_cached = lru_cache(maxsize=cache_size)(self._translate_single)
    
    def _translate_single(self, text: str) -> str:
        if not text or not text.strip():
            return ""
        try:
            result = self.translator.translate(text.strip())
            return result if result else text
        except Exception as e:
            logger.error(f"Translation failed: {str(e)}")
            raise
    
    def translate(self, text: str, fallback_to_original: bool = True) -> str:
        if not text or not text.strip():
            return ""
        try:
            return self._translate_cached(text.strip())
        except Exception as e:
            if fallback_to_original:
                logger.warning(f"Translation failed, using original: {str(e)}")
                return text
            raise


# ============================================================================
# SCRIPT-SPECIFIC PROCESSING
# ============================================================================

def transliterate_romanized_nepali(text: str) -> str:
    """Transliterate Romanized Nepali to Devanagari using ITRANS"""
    if not TRANSLITERATION_AVAILABLE:
        logger.warning("Transliteration not available")
        return text
    
    try:
        result = transliterate(text, sanscript.ITRANS, sanscript.DEVANAGARI)
        return result if result else text
    except Exception as e:
        logger.warning(f"Transliteration failed: {e}")
        return text


def translate_latin_spans(text: str, translator: CachedNepaliTranslator) -> str:
    """Translate ONLY Latin word spans inside a Devanagari-dominant sentence"""
    def repl(match):
        latin_text = match.group(0)
        translated = translator.translate(latin_text, fallback_to_original=True)
        return f" {translated} "
    
    return re.sub(r"[A-Za-z][A-Za-z\s]{2,}", repl, text)


# ============================================================================
# EMOJI PROCESSING (APPLIED AFTER SCRIPT CONVERSION)
# ============================================================================

def replace_emojis_semantic(text: str, preserve_spacing: bool = True) -> str:
    """Replace emojis with Nepali text"""
    for emoji_char, nepali_text in EMOJI_TO_NEPALI.items():
        if preserve_spacing:
            text = text.replace(emoji_char, f" {nepali_text} ")
        else:
            text = text.replace(emoji_char, nepali_text)
    
    text = emoji.replace_emoji(text, replace=" ")
    return text


# ============================================================================
# NORMALIZATION
# ============================================================================

DIRGHIKARAN_MAP = {
    "\u200d": "",
    "\u200c": "",
    "à¥¤": ".",
    "à¥¥": ".",
}


def normalize_devanagari(text: str) -> str:
    """Normalize Devanagari-specific characters"""
    for k, v in DIRGHIKARAN_MAP.items():
        text = text.replace(k, v)
    return text


def clean_text_basic(text: str) -> str:
    """Basic cleaning: URLs, mentions, whitespace"""
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#(\w+)", r"\1", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


# ============================================================================
# MAIN PREPROCESSING PIPELINE
# ============================================================================

class HateSpeechPreprocessor:
    """
    Unified preprocessing pipeline for Nepali hate speech classification
    WITH EMOJI INTELLIGENCE
    """

    def __init__(
        self,
        model_type: Literal["xlmr", "mbert", "nepalibert"] = "xlmr",
        translate_english: bool = True,
        cache_size: int = 2000,
        use_emoji_tags: bool = False
    ):
        self.model_type = model_type
        self.translate_english = translate_english
        self.use_emoji_tags = use_emoji_tags
        self.translator = CachedNepaliTranslator(cache_size) if translate_english else None

    def preprocess(self, text: str, verbose: bool = False) -> Tuple[str, Dict[str, int]]:
        """
        Main preprocessing pipeline - ALL outputs in Devanagari + emoji features
        
        Returns:
            Tuple of (preprocessed_text, emoji_features)
        """
        if not isinstance(text, str) or not text.strip():
            return "", {
                'has_hate_emoji': 0, 'has_mockery_emoji': 0, 'has_positive_emoji': 0,
                'has_sadness_emoji': 0, 'has_fear_emoji': 0, 'has_disgust_emoji': 0,
                'hate_emoji_count': 0, 'mockery_emoji_count': 0, 'positive_emoji_count': 0,
                'sadness_emoji_count': 0, 'fear_emoji_count': 0, 'disgust_emoji_count': 0,
                'total_emoji_count': 0, 'hate_to_positive_ratio': 0.0, 'has_mixed_sentiment': 0
            }

        original_text = text
        
        if verbose:
            print(f"ğŸ“¥ Original: {original_text}")

        # STEP 1: EXTRACT EMOJI FEATURES (BEFORE any processing)
        emoji_features = extract_emoji_features(original_text)
        
        if verbose:
            print(f"ğŸ˜Š Emoji Features: Hate={emoji_features['hate_emoji_count']}, "
                  f"Positive={emoji_features['positive_emoji_count']}, "
                  f"Total={emoji_features['total_emoji_count']}")

        # STEP 2: DETECT SCRIPT TYPE
        script_type, detection_details = detect_script_type(text)
        
        if verbose:
            print(f"ğŸ” Detected Script: {script_type.upper()} "
                  f"(confidence: {detection_details.get('confidence', 0):.2%})")

        # STEP 3: BASIC CLEANING
        text = clean_text_basic(text)

        # STEP 4: SCRIPT-SPECIFIC PROCESSING
        if script_type == "devanagari":
            processed = text
            if self.translate_english:
                processed = translate_latin_spans(processed, self.translator)
        
        elif script_type == "romanized_nepali":
            processed = transliterate_romanized_nepali(text)
        
        elif script_type == "english":
            if self.translate_english:
                processed = self.translator.translate(text, fallback_to_original=True)
            else:
                processed = text
        
        elif script_type == "mixed":
            processed = transliterate_romanized_nepali(text)
            if self.translate_english:
                processed = translate_latin_spans(processed, self.translator)
        else:
            processed = text

        # STEP 5: EMOJI REPLACEMENT
        processed = replace_emojis_semantic(processed)

        # STEP 6: NORMALIZATION
        final = normalize_devanagari(processed)
        final = re.sub(r"\s+", " ", final).strip()

        return final, emoji_features


# ============================ MODEL LOADING ================================
TEST_FILE = '/kaggle/input/nepalihatee/test_processed.json'
LOCAL_MODEL_PATH = '/kaggle/input/xlm-roberta-nepali-hate'
HF_MODEL_ID = "UDHOV/xlm-roberta-large-nepali-hate-classification"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_model():
    if os.path.exists(LOCAL_MODEL_PATH):
        tokenizer = AutoTokenizer.from_pretrained(LOCAL_MODEL_PATH)
        model = AutoModelForSequenceClassification.from_pretrained(LOCAL_MODEL_PATH)
        le = joblib.load(os.path.join(LOCAL_MODEL_PATH, "label_encoder.pkl"))
        print("âœ… Model loaded locally")
    else:
        tokenizer = AutoTokenizer.from_pretrained(HF_MODEL_ID)
        model = AutoModelForSequenceClassification.from_pretrained(HF_MODEL_ID)
        le_path = hf_hub_download(HF_MODEL_ID, "label_encoder.pkl")
        le = joblib.load(le_path)
        print("âœ… Model loaded from HuggingFace")
    model.to(DEVICE).eval()
    return model, tokenizer, le


# ============================ FONT HANDLING ==================================
def apply_nepali_font_and_emoji(ax: plt.Axes, nepali_font: Optional[FontProperties] = None,
                                 texts: Optional[list] = None, is_tick_labels: bool = True) -> None:
    """
    Apply Nepali font to Devanagari text while preserving emojis
    
    CRITICAL: Emojis render correctly without special font handling in matplotlib
    We only need to apply Nepali font to Devanagari characters
    """
    if nepali_font is None:
        return
    
    if is_tick_labels or texts is None:
        for txt in ax.get_yticklabels():
            text_content = txt.get_text()
            # Only apply font if there's Devanagari content
            if regex.search(r'\p{Devanagari}', text_content):
                txt.set_fontproperties(nepali_font)
                txt.set_fontsize(11)
    else:
        for txt in texts:
            text_content = txt.get_text()
            if regex.search(r'\p{Devanagari}', text_content):
                txt.set_fontproperties(nepali_font)


def create_display_text_with_emojis(original_text: str, preprocessed_text: str) -> Tuple[List[str], List[str]]:
    """
    Create aligned display tokens that preserve original emojis for visualization
    while using preprocessed text for model predictions
    
    Strategy:
    1. Extract emojis from original text with their positions
    2. Create word-level alignment between original and preprocessed
    3. For display: use original tokens (with emojis)
    4. For model: use preprocessed tokens (emojis replaced with Nepali)
    
    Returns:
        Tuple of (display_tokens, model_tokens)
    """
    # Extract emojis and their positions from original
    original_tokens = original_text.split()
    preprocessed_tokens = preprocessed_text.split()
    
    display_tokens = []
    model_tokens = []
    
    orig_idx = 0
    proc_idx = 0
    
    while orig_idx < len(original_tokens):
        orig_token = original_tokens[orig_idx]
        
        # Check if this token contains emoji
        has_emoji = any(c in emoji.EMOJI_DATA for c in orig_token)
        
        if has_emoji:
            # For display: keep original emoji
            display_tokens.append(orig_token)
            
            # For model: use the Nepali translation (could span multiple preprocessed tokens)
            # Find corresponding preprocessed tokens
            if proc_idx < len(preprocessed_tokens):
                # If emoji was standalone, it maps to one Nepali word
                # If mixed with text, it might map to multiple words
                model_token_parts = []
                
                # Simple heuristic: if original token is pure emoji, take next preprocessed token
                # Otherwise, take multiple preprocessed tokens
                if all(c in emoji.EMOJI_DATA or c.isspace() for c in orig_token):
                    # Pure emoji token
                    if proc_idx < len(preprocessed_tokens):
                        model_token_parts.append(preprocessed_tokens[proc_idx])
                        proc_idx += 1
                else:
                    # Mixed token - take corresponding preprocessed tokens
                    # This is approximate - in practice, emojis are usually standalone
                    if proc_idx < len(preprocessed_tokens):
                        model_token_parts.append(preprocessed_tokens[proc_idx])
                        proc_idx += 1
                
                model_tokens.append(" ".join(model_token_parts) if model_token_parts else orig_token)
            else:
                model_tokens.append(orig_token)
        else:
            # No emoji: use preprocessed version for both
            if proc_idx < len(preprocessed_tokens):
                display_tokens.append(preprocessed_tokens[proc_idx])
                model_tokens.append(preprocessed_tokens[proc_idx])
                proc_idx += 1
            else:
                display_tokens.append(orig_token)
                model_tokens.append(orig_token)
        
        orig_idx += 1
    
    # Handle any remaining preprocessed tokens
    while proc_idx < len(preprocessed_tokens):
        token = preprocessed_tokens[proc_idx]
        display_tokens.append(token)
        model_tokens.append(token)
        proc_idx += 1
    
    return display_tokens, model_tokens


# ============================ MODEL WRAPPER ================================
class XLMRobertaExplainer:
    def __init__(self, model, tokenizer, label_encoder, preprocessor):
        self.model = model
        self.tokenizer = tokenizer
        self.class_names = label_encoder.classes_.tolist()
        self.preprocessor = preprocessor

    def preprocess_text(self, text: str) -> Tuple[str, Dict[str, int]]:
        """Preprocess text using the HateSpeechPreprocessor"""
        return self.preprocessor.preprocess(text, verbose=False)

    def predict_proba(self, texts):
        """Predict probabilities for texts (expects preprocessed text)"""
        if isinstance(texts, str):
            texts = [texts]
        
        enc = self.tokenizer(
            texts, padding=True, truncation=True, max_length=256, return_tensors="pt"
        ).to(DEVICE)
        
        with torch.no_grad():
            probs = torch.softmax(self.model(**enc).logits, dim=-1)
        
        return probs.cpu().numpy()

    def predict_with_analysis(self, text: str):
        """Predict with full analysis including preprocessing info"""
        # Preprocess
        preprocessed, emoji_features = self.preprocess_text(text)
        
        # Predict
        probs = self.predict_proba(preprocessed)[0]
        pred_idx = int(np.argmax(probs))
        
        return {
            "original_text": text,
            "preprocessed_text": preprocessed,
            "emoji_features": emoji_features,
            "predicted_label": self.class_names[pred_idx],
            "confidence": float(probs[pred_idx]),
            "probabilities": {label: float(prob) for label, prob in zip(self.class_names, probs)}
        }


# ============================ LIME / SHAP CLASSES ==========================
class LIMEExplainer:
    def __init__(self, model_explainer, nepali_font=None):
        self.model_explainer = model_explainer
        self.nepali_font = nepali_font
        self.explainer = LimeTextExplainer(
            class_names=model_explainer.class_names, random_state=42
        )

    def explain_and_visualize(self, original_text: str, preprocessed_text: str):
        """
        Explain predictions with LIME
        
        Args:
            original_text: Original text with emojis
            preprocessed_text: Preprocessed text for model
        """
        # Get LIME explanation using preprocessed text
        exp = self.explainer.explain_instance(
            preprocessed_text, 
            self.model_explainer.predict_proba, 
            num_samples=200
        )
        
        # Get token weights from LIME
        token_weights = dict(exp.as_list())
        
        # Create aligned tokens for display
        display_tokens, model_tokens = create_display_text_with_emojis(
            original_text, preprocessed_text
        )
        
        # Map weights to display tokens
        word_scores = []
        for display_tok, model_tok in zip(display_tokens, model_tokens):
            # Find score for this token in LIME results
            score = 0.0
            for lime_token, weight in token_weights.items():
                if lime_token in model_tok or model_tok in lime_token:
                    score += weight
            word_scores.append((display_tok, score))
        
        if not word_scores:
            print("âš ï¸ No explainable words found")
            return
        
        # Visualize
        features, weights = zip(*word_scores)
        y_pos = range(len(features))
        
        fig, ax = plt.subplots(figsize=(10, max(6, len(features) * 0.4)))
        colors = ['red' if w < 0 else 'green' for w in weights]
        ax.barh(y_pos, weights, color=colors, alpha=0.6)
        ax.set_yticks(y_pos)
        ax.set_yticklabels(features, fontsize=12)
        ax.invert_yaxis()
        ax.set_xlabel("Contribution to Prediction", fontsize=12)
        ax.set_title("LIME Feature Importance (Red=Against, Green=For)", fontsize=14)
        ax.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
        
        # Apply Nepali font to Devanagari text (emojis render naturally)
        apply_nepali_font_and_emoji(ax, self.nepali_font)
        
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, f"lime_explanation_{abs(hash(original_text)) % 10**8}.png")
        plt.savefig(out_path, dpi=300, bbox_inches="tight")
        plt.show()
        print(f"âœ… LIME image saved to: {out_path}")

class SHAPExplainer:
    def __init__(self, model_explainer, nepali_font=None):
        self.model_explainer = model_explainer
        self.nepali_font = nepali_font
        # Don't initialize explainer here - we'll use a simpler approach

    def explain_and_visualize(self, original_text: str, preprocessed_text: str):
        """
        Explain predictions with SHAP using a simplified masking approach
        
        Since SHAP's text masker doesn't work well with our preprocessing pipeline,
        we use a simpler word-level masking approach.
        
        Args:
            original_text: Original text with emojis
            preprocessed_text: Preprocessed text for model
        """
        try:
            # Use a simpler masking approach compatible with preprocessing
            from shap import Explainer, maskers
            
            # Create a custom prediction function that handles masked text
            def predict_masked(masked_texts):
                """Predict on masked texts - handles both strings and arrays"""
                if isinstance(masked_texts, np.ndarray):
                    # Convert masked arrays to text
                    if masked_texts.ndim == 1:
                        texts = [' '.join(str(t) for t in masked_texts if str(t).strip())]
                    else:
                        texts = [' '.join(str(t) for t in row if str(t).strip()) for row in masked_texts]
                elif isinstance(masked_texts, str):
                    texts = [masked_texts]
                elif isinstance(masked_texts, list):
                    texts = masked_texts
                else:
                    texts = [str(masked_texts)]
                
                return self.model_explainer.predict_proba(texts)
            
            # Create explainer with text masker
            explainer = Explainer(predict_masked, maskers.Text(preprocessed_text))
            
            # Get SHAP values
            sv = explainer([preprocessed_text])[0]
            shap_tokens = list(sv.data)
            values_array = np.array(sv.values)
            
        except Exception as e:
            print(f"âš ï¸ SHAP explainer failed: {e}")
            print("ğŸ“Š Using alternative token-level attribution method...")
            
            # Fallback: Use gradient-based attribution
            shap_tokens, values_array = self._gradient_based_attribution(
                preprocessed_text
            )
        
        # Get predicted class
        pred_probs = self.model_explainer.predict_proba([preprocessed_text])[0]
        class_idx = int(np.argmax(pred_probs))
        
        # Extract values for predicted class
        if values_array.ndim == 1:
            token_values = values_array
        elif values_array.ndim == 2:
            token_values = values_array[:, class_idx]
        elif values_array.ndim == 3:
            token_values = values_array[0, :, class_idx]
        else:
            token_values = values_array.flatten()[:len(shap_tokens)]
        
        # Create aligned tokens for display
        display_tokens, model_tokens = create_display_text_with_emojis(
            original_text, preprocessed_text
        )
        
        # Map SHAP values to display tokens
        word_scores = self._align_shap_values(
            display_tokens, model_tokens, shap_tokens, token_values
        )
        
        if not word_scores:
            print("âš ï¸ No explainable words found")
            return
        
        # Visualize with highlighted text
        self._visualize_shap(word_scores, class_idx, original_text)
    
    def _gradient_based_attribution(self, text: str) -> Tuple[List[str], np.ndarray]:
        """
        Fallback: Simple word-level attribution using occlusion
        
        This method masks each word and measures the change in prediction
        """
        words = text.split()
        base_probs = self.model_explainer.predict_proba([text])[0]
        base_pred_idx = int(np.argmax(base_probs))
        base_score = base_probs[base_pred_idx]
        
        attributions = []
        for i in range(len(words)):
            # Mask the word
            masked_words = words[:i] + words[i+1:]
            masked_text = ' '.join(masked_words)
            
            if not masked_text.strip():
                attributions.append(base_score)
                continue
            
            # Get prediction without this word
            masked_probs = self.model_explainer.predict_proba([masked_text])[0]
            masked_score = masked_probs[base_pred_idx]
            
            # Attribution = how much score drops when word is removed
            attribution = base_score - masked_score
            attributions.append(attribution)
        
        return words, np.array(attributions)
    
    def _align_shap_values(self, display_tokens: List[str], model_tokens: List[str],
                          shap_tokens: List[str], token_values: np.ndarray) -> List[Tuple[str, float]]:
        """Align SHAP values with display tokens"""
        word_scores = []
        
        # Simple alignment: assume one-to-one or aggregate subwords
        if len(display_tokens) == len(model_tokens):
            # Direct alignment
            for i, (display_tok, model_tok) in enumerate(zip(display_tokens, model_tokens)):
                # Find matching SHAP tokens
                score = 0.0
                for j, shap_tok in enumerate(shap_tokens):
                    if j < len(token_values) and (shap_tok in model_tok or model_tok in shap_tok):
                        score += float(token_values[j])
                word_scores.append((display_tok, score))
        else:
            # Fallback: distribute values evenly
            for display_tok in display_tokens:
                score = np.mean(token_values) if len(token_values) > 0 else 0.0
                word_scores.append((display_tok, score))
        
        return word_scores
    
    def _visualize_shap(self, word_scores: List[Tuple[str, float]], 
                       class_idx: int, original_text: str):
        """Create SHAP visualization with emojis"""
        if not word_scores:
            return
        
        max_val = max(abs(v) for _, v in word_scores) + 1e-6
        
        fig, ax = plt.subplots(figsize=(max(10, 0.5 * len(word_scores)), 3))
        ax.axis("off")
        
        x, y = 0.01, 0.5
        text_objs = []
        
        for word, val in word_scores:
            # Color intensity based on absolute value
            intensity = min(abs(val) / max_val, 1.0)
            
            # Red for negative (against prediction), Green for positive (for prediction)
            if val < 0:
                color = (1.0, 1.0 - intensity * 0.7, 1.0 - intensity * 0.7)  # Red scale
            else:
                color = (1.0 - intensity * 0.7, 1.0, 1.0 - intensity * 0.7)  # Green scale
            
            txt = ax.text(
                x, y, f" {word} ",
                fontsize=14,
                bbox=dict(
                    facecolor=color, 
                    edgecolor='gray', 
                    alpha=0.8, 
                    boxstyle="round,pad=0.4"
                )
            )
            text_objs.append(txt)
            
            # Update x position (emojis take less space visually)
            char_width = 0.025 if any(c in emoji.EMOJI_DATA for c in word) else 0.04
            x += char_width * len(word) + 0.01
            
            if x > 0.92:
                x = 0.01
                y -= 0.35
        
        # Apply Nepali font to Devanagari text (emojis render naturally)
        apply_nepali_font_and_emoji(ax, self.nepali_font, texts=text_objs, is_tick_labels=False)
        
        class_name = self.model_explainer.class_names[class_idx]
        ax.text(0.5, 0.95, f"SHAP Explanation (Predicted: {class_name})", 
                ha='center', va='top', fontsize=14, fontweight='bold',
                transform=ax.transAxes)
        
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, f"shap_explanation_{class_name}_{abs(hash(original_text)) % 10**8}.png")
        plt.savefig(out_path, dpi=300, bbox_inches="tight")
        plt.show()
        print(f"âœ… SHAP image saved to: {out_path}")

# ============================ EXECUTION =====================================
if __name__ == "__main__":
    # Load model
    model, tokenizer, le = load_model()
    
    # Initialize preprocessor
    preprocessor = HateSpeechPreprocessor(
        model_type="xlmr",
        translate_english=False,
        cache_size=2000
    )
    
    # Initialize explainer
    explainer = XLMRobertaExplainer(model, tokenizer, le, preprocessor)
    
    # Load test data
    df = pd.read_json(TEST_FILE)
    pd.set_option('display.max_colwidth', None)
    
    SAMPLES_PER_CLASS = 3
    sampled_rows = []
    
    # Sample data ensuring emoji representation
    for label, group in df.groupby("Label_Multiclass"):
        # Prioritize rows with emojis
        has_emoji = group[group["Comment"].apply(
            lambda x: any(c in emoji.EMOJI_DATA for c in str(x))
        )]
        no_emoji = group.drop(has_emoji.index)
        
        selected = []
        
        # Pick one with emoji if available
        if len(has_emoji) > 0:
            selected.append(has_emoji.sample(n=1, random_state=42))
        
        # Fill remaining samples
        remaining_needed = SAMPLES_PER_CLASS - len(selected)
        if remaining_needed > 0:
            pool = no_emoji if len(no_emoji) >= remaining_needed else group
            selected.append(pool.sample(n=min(len(pool), remaining_needed), random_state=42))
        
        if selected:
            sampled_rows.append(pd.concat(selected))
    
    sampled_df = pd.concat(sampled_rows) if sampled_rows else df.head(5)
    
    # Process samples
    for i, (_, row) in enumerate(sampled_df.iterrows(), start=1):
        original_text = str(row["Comment"])
        true_label = row.get("Label_Multiclass", "N/A")
        
        print(f"\n{'='*80}")
        print(f"Sample {i}/{len(sampled_df)}")
        print(f"{'='*80}")
        
        # Analyze
        analysis = explainer.predict_with_analysis(original_text)
        
        print(f"\nğŸ“ Original Text: {analysis['original_text']}")
        print(f"ğŸ”„ Preprocessed: {analysis['preprocessed_text']}")
        print(f"\nğŸ˜Š Emoji Features:")
        for key, val in analysis['emoji_features'].items():
            if val > 0 or 'count' in key:
                print(f"   {key}: {val}")
        print(f"\nğŸ¯ True Label: {true_label}")
        print(f"ğŸ¤– Predicted: {analysis['predicted_label']}")
        print(f"ğŸ“Š Confidence: {analysis['confidence']:.4f}")
        print(f"\nğŸ“ˆ All Probabilities:")
        for label, prob in analysis['probabilities'].items():
            print(f"   {label}: {prob:.4f}")
        
        # LIME Explanation
        if LIME_AVAILABLE:
            print(f"\n{'â”€'*80}")
            print("ğŸ” LIME Explanation")
            print(f"{'â”€'*80}")
            LIMEExplainer(explainer, NEPALI_FONT).explain_and_visualize(
                analysis['original_text'],
                analysis['preprocessed_text']
            )
        
        # SHAP Explanation
        if SHAP_AVAILABLE:
            print(f"\n{'â”€'*80}")
            print("ğŸ” SHAP Explanation")
            print(f"{'â”€'*80}")
            SHAPExplainer(explainer, NEPALI_FONT).explain_and_visualize(
                analysis['original_text'],
                analysis['preprocessed_text']
            )
        
        print(f"\n{'='*80}\n")
    
    print(f"\nâœ… All explanations completed!")
    print(f"ğŸ“ Outputs saved to: {OUT_DIR}")
```
